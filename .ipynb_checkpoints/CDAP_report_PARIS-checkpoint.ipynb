## Forward problem

### Mathematical background

, which allows to model the displacement in 3 directions of any location in space that is generated by a source of polygonal shape (with any number of vertices) and of constant load. 
For a given source and location, the three components of the displacement can be computed with the following equations: 

\begin{equation}
\begin{pmatrix}
    dx \\
    dy
\end{pmatrix}  =  \frac{p_z}{4 \pi \mu } \sum_{i=1}^n  \begin{pmatrix}
    y_{i+1} - y_i \\
    x_i - x_{i+1}
\end{pmatrix} \left(  \frac{\mu}{\lambda + \mu}  S_{1i} + S_{2i}  \right) 
\label{eq:dxdy}
\end{equation}

\begin{equation}
dz =  \frac{p_z}{4 \pi \mu } \sum_{i=1}^n (x_i y_{i+1} - x_{i+1} y_i ) \{  S_{2i} - \frac{\mu}{\lambda + \mu} \left[ S_{3i} - z \alpha(\bf{0})  \right]  \} 
\label{eq:dz}
\end{equation}

Where $\mu$ and $\lambda$ are the Lame parameters, $p_z$ is the load, $x$ and $y$ are the coordinates of the vertices of the load,  ...  Because the relationship between the load $p_z$ and the displacements is linear, the displacements in 3D at all measurement location $\bf{d}$ from the contribution of discrete sources of constant load $\bf{p}$ can be expressed as 

\begin{equation}
    \bf{G} \bf{p} = \bf{d} 
\end{equation}

Where G is a matrix of size [N,M], where N is the number of measurements (3times the number of mesurement locations) and M is the number of discrte sources. This linear system is at the core of the inversion process detailed in the following section.


### Implementation

Implementation to compute the displacement a 1 location generated by 1 source(need to loop over the function to compute the displacements at different locations and for different sources)

```python
from scipy.integrate import quad
from numpy import arctan, tan, cos, sin, arccos, arcsin, sqrt, log, log10, pi

def compute_I1(r, z):
    nv = r.shape[0]
    rot = np.array([[0, 1], [-1, 0]])
    I1 = np.array([0, 0])

    for i in range(nv):
        ri = r[i, :]
        if i == nv - 1:
            rj = r[0, :]
        else:
            rj = r[i + 1, :]
        rjio = np.dot(rot, (rj - ri))
        a = np.dot(rj-ri,rj-ri)
        b = np.dot(ri,rj-ri)
        c = np.dot(ri,ri)
        d = c + z**2
        e = d - b**2 / a
        t0 = b / a
        t1 = 1 + b / a
        if np.linalg.norm(rjio) != 0:
            S1i = compute_S1i(a, e, z, t0, t1)
            I1 = I1 + np.dot(S1i, rjio)
    return I1

def compute_I2(r, z):
    nv = r.shape[0]
    rot = np.array([[0, 1], [-1, 0]])
    I2 = 0
    for i in range(nv):
        ri = r[i, :]
        if i == nv - 1:
            rj = r[0, :]
        else:
            rj = r[i + 1, :]
        rjio = np.dot(rot, (rj - ri))
        a = np.dot(rj - ri, rj - ri)
        b = np.dot(ri, rj - ri)
        c = np.dot(ri, ri)
        d = c + z**2
        e = d - b**2 / a
        t0 = b / a
        t1 = 1 + b / a
        if np.linalg.norm(rjio) != 0:
            S2i = compute_S2i(a, e, t0, t1)
            I2 = I2 + np.dot(rjio, S2i)

    return I2

def compute_I3(r, z):
    nv = r.shape[0]
    rot = np.array([[0, 1], [-1, 0]])
    I3 = 0
    for i in range(nv):
        ri = r[i, :]
        if i == nv - 1:
            rj = r[0, :]
        else:
            rj = r[i + 1, :]
        rjo = np.dot(ri, np.dot(rot, rj))
        a = np.dot(rj - ri, rj - ri)
        b = np.dot(ri, rj - ri)
        c = np.dot(ri, ri)
        d = c + z**2
        e = d - b**2 / a
        t0 = b / a
        t1 = 1 + b / a
        if rjo != 0:
            S2i = compute_S2i(a, e, t0, t1)
            I3 = I3 + rjo * S2i
    return I3

def compute_I4(r, z):
    nv = r.shape[0]
    rot = np.array([[0, 1], [-1, 0]])
    I4 = 0
    for i in range(nv):
        ri = r[i, :]
        if i == nv - 1:
            rj = r[0, :]
        else:
            rj = r[i + 1, :]
        rjo = np.dot(ri, np.dot(rot, rj))
        a = np.dot(rj - ri, rj - ri)
        b = np.dot(ri, rj - ri)
        c = np.dot(ri, ri)
        d = c + z**2
        A = c / a - b**2 / a**2
        B = d / a - b**2 / a**2
        t0 = b / a
        t1 = 1 + b / a

        if rjo != 0:
            S3i = compute_S3i(a, A, B, t0, t1)
            I4 = I4 + rjo * S3i
    return I4

def compute_S1i(a, e, z, t0, t1):
    S1i = sqrt(a) * t0 - sqrt(a) * t1
    if e - z**2 != 0:
        S1i = S1i - sqrt(e - z**2) * arctan(sqrt(a) * t0 / sqrt(e - z**2)) + \
              sqrt(e - z**2) * arctan(sqrt(a) * t1 / sqrt(e - z**2)) + \
              sqrt(e - z**2) * arctan(sqrt(a) * t0 * z / (sqrt(e + a * t0**2) * sqrt(e - z**2))) - \
              sqrt(e - z**2) * arctan(sqrt(a) * t1 * z / (sqrt(e + a * t1**2) * sqrt(e - z**2)))
    if z != 0:
        S1i = S1i - z * np.log(a * t0 + np.sqrt(a) * np.sqrt(e + a * t0**2)) + \
              z * log(a * t1 + sqrt(a) * sqrt(e + a * t1**2))
    if t0 != 0:
        S1i = S1i - sqrt(a) * t0 * log(sqrt(e + a * t0**2) + z)
    if t1 != 0:
        S1i = S1i + sqrt(a) * t1 * log(sqrt(e + a * t1**2) + z)
    return S1i / sqrt(a)

def compute_S2i(a, e, t0, t1):
    S2i = log((a * t1 + np.sqrt(a * (a * t1**2 + e))) / (a * t0 + np.sqrt(a * (a * t0**2 + e)))) / sqrt(a)
    return S2i

def compute_S3i(a, A, B, t0, t1):
    S3i = arctan(t1 * sqrt((B - A) / (A * (B + t1**2)))) - arctan(t0 * sqrt((B - A) / (A * (B + t0**2))))
    S3i = S3i * sqrt((B - A) / A) + log((t1 + sqrt(B + t1**2)) / (t0 + sqrt(B + t0**2))) / sqrt(a)
    return S3i

def alpha(r):
    a = 0
    for i in range(len(r)):
        ri = r[i, :]
        if i < len(r) - 1:
            rj = r[i + 1, :]
        else:
            rj = r[0, :]
        if np.linalg.norm(rj) > 0 and np.linalg.norm(ri) > 0:
            aj = np.arctan2(rj[1], rj[0])
            ai = np.arctan2(ri[1], ri[0])
            da = aj - ai
            if da > np.pi:
                da = da - 2 * np.pi
            if da < -np.pi:
                da = da + 2 * np.pi
            if np.abs(da) == np.pi:
                da = 0
            a = a + da
    return a

def load2disp(xyz, r, p, E, v):
    """This function computes the displacement of a given point at coordinates xyz for a given source of vertices r which is a polygon of any shape 
      
    The required shape for xyz is (3,) and the required size for r is (n,2). 
    For a square source has a shape (4,2): [[x_bot_left,y_bot_left],
                                           [x_bot_right],[y_bot_right],
                                           [x_top_right],[y_top_right]
                                           [x_top_left] ,[y_top_left]] -> anti clockwise order"""
    # Translating the coordinates of the vertices so that the considered point is located at the middle
    r_translated = np.zeros(r.shape)
    r_translated[:,0] = -r[:,0] + xyz[0] 
    r_translated[:,1] = -r[:,1] + xyz[1]   
    z = xyz[2]
    # Lame constants
    l = E * v / ((1 + v) * (1 - 2 * v))
    m = E / (2 * (1 + v))
    # Integrals
    I1 = compute_I1(r_translated, z)
    I3 = compute_I3(r_translated, z)
    I4 = compute_I4(r_translated, z)
    # Displacements
    U = np.zeros((3, 1))
    U[0:2, 0] = -I1*p/(4*pi*(l+m))
    U[2, 0] = (I3*p/(4*pi*m)) + (I4 * p / (4 * pi * (l + m)))
    if z != 0:
        I2 = compute_I2(r_translated, z)
        al = alpha(r_translated)
        U[0:2, 0] -= I2 * z * p / (4 * np.pi * m)
        U[2, 0] -= z * al * p / (4 * np.pi * (l + m))
    return U

def create_source_mesh(x0,x1,y0,y1,ps):
    """This function creates a 2D mesh for which each cell is associated with a source and contains the coordinates of the 4 vertices of the 
    rectangular source.
    :input: x0 = smallest x value, x1 = largest x value, y0 = smallest y value, y1 = largest y value
    :output: numpy array of shape (ny,nx,4,2)""""
    vertice_mesh = np.zeros((ps.shape[0],ps.shape[1],4,2))
    x_num = ps.shape[1]
    y_num = ps.shape[0]
        
    xv = np.linspace(x0,x1,x_num+1)
    yv = np.linspace(y0,y1,y_num+1)
    
    for i in range(x_num):
        for j in range(y_num):
            vertice_mesh[j,i,:,:] = np.array([[xv[i],yv[j]],
                                             [xv[i+1],yv[j]],
                                             [xv[i+1],yv[j+1]],
                                             [xv[i],yv[j+1]]])
    return vertice_mesh
```

Implementation to build the G matrix to perform efficient matrix vector products (G can be saved to be used for multiple runs).

```python
def build_G(rs,xs,ys,l,m):
    """Fonction pour build G, la matrice qui relie les pressions au displacements
    :input: rs = source mesh obtained with create mesh
            xs, ys, = x and y locaiton of the station, y is assumed to be 0 (displacement recorded at the surface)
            l,m = Lame parameters"""
    source_number = len(rs[:,0])*len(rs[0,:]) 
    data_number = len(xs)*3
    G = np.zeros((data_number,source_number))
    rs_formatted = rs.reshape(source_number,4,2) 
    for i in range(source_number): # on remplit à présent la matrice G et pour ça on a besoin d'itérer sur les sources        
        r = rs_formatted[i] 
        for j in range(len(xs)): #on doit aussi boucler sur les positions de la station; 1 station fournit 3 data
            xyz = [xs[j],ys[j],0] #z = 0 on présume car c'est des enregistrements à la surface et qu'il accepte pas des altitutdes positives 
            # Translating the coordinates of the vertices so that the considered point is located at the middle
            r_translated = np.zeros(r.shape)
            r_translated[:,0] = -r[:,0] + xyz[0] 
            r_translated[:,1] = -r[:,1] + xyz[1]   
            z = xyz[2]

            # Integrals
            I1 = compute_I1(r_translated, z)
            I3 = compute_I3(r_translated, z)
            I4 = compute_I4(r_translated, z)

            # Displacements
            tmp = -I1/(4*np.pi*(l+m)) #coefficient pour la source i et pour la station j pour les déplacements horizontaux
            alpha = tmp[0]
            beta  = tmp[1]
            gamma = I3/(4*np.pi*m) + I4/(4*np.pi*(l + m)) #coefficent pour la source i  et la station j pour les déplacements verticaux
            
            starting_idx = j*3 #car 3 données par station
            G[starting_idx,i] = alpha
            G[starting_idx+1,i] = beta
            G[starting_idx+2,i] = gamma
    return G
```

For our problem we are in the situation of an overdetermined problem, asCalifornia has a lot of GPS stations 

For an overdtermined problem, the Least-square solution writes of our problem writes  

\begin{equation}
(\bf{G^T}\bf{G})^{-1} \bf{G^T} \bf{d} = \bf{p}
    \label{eq:LSQ_overdetermined}
\end{equation}


However, theGPS stations are not spread uniformly across California therefore due to noise the inversion is very unstable, therefore some physics must be added through regularization. Different regulariation were investigated to add the physics : 


METTRE LES EQUATIONS DU PROBLEM AVEC CES DEUX APPROCHES 

- Laplacian regularization -> penalty on the gradient of m, to ensures that the neighbours loads are not too different. 

- Total variation  

- Gaussian Regularization : 


\begin{equation}
bla = bla
\end{equation}


total variation is described in \cite{Mallat}



In these expressions the value of $\lambda$ is theorically obtained through L-curve, where one plots the norm of the misfit as afucntion of the norm of the regularizer. In this case the value of $\lambda$ is taken as the value which leads to the best trade-off between fitting the best the data and minimzing the most the regularizer (inthis case  gradient of the load distribution = smoothing), which is located in the elbow of the L-curve.

We however found out that the L-curve tends to underestimate the value of lambda when the noise is greater than the actual displacement, which is likely during Hilary Hurricane. We found an empirical criterion which appears to be able to consistently predict the correct value of lambda which such high noise level. This criterion is the average of the absolute value of the distance between the recorded horizontal displacements and the displacements computed using the inverted load distriubtion.

```python
def kfromij(i,j,nx): # fonction pour trouver les indices globaux des sources car les m sont regroupés dans un vecteur 
    return j*nx+i #numerotation globale pour crééer le laplacien

def build_laplacian(rs):
    """Fonction qui compute le laplacien par différence finie servant pour la régularization
    :input: rs = mesh created with create mesh"""
    shape = rs[:,:,0,0].shape
    dx = (rs[-1,-1,2,0]-rs[0,0,0,0])/len(rs[0,:,0,0])
    dy = (rs[-1,-1,2,1]-rs[0,0,0,1])/len(rs[:,0,0,0])
    source_number = shape[0]*shape[1]
    nx, ny = shape[1], shape[0]
    laplacien = np.zeros((source_number,source_number))
    for i in np.arange(0,nx):
        for j in np.arange(0,ny):
                k=kfromij(i,j,nx)
                laplacien[k,k]=-2/dx**2-2/dy**2 #central term
                #left term
                if i >0 :
                    kl=kfromij(i-1,j,nx)
                    laplacien[k,kl]=1/dx**2
                #right term
                if i < nx-1 :
                    kr=kfromij(i+1,j,nx)
                    laplacien[k,kr]=1/dx**2
                #bottom term
                if j >0 :
                    kb=kfromij(i,j-1,nx)
                    laplacien[k,kb]=1/dy**2
                #top term
                if j < ny-1 :
                    kt=kfromij(i,j+1,nx)
                    laplacien[k,kt]=1/dy**2
    return laplacien     

def build_gaussian_inv(rs, sigma, normalized=False):
    """Builds a Gaussian matrix for regularization and returns its inverse to remain consistant with the Laplacian implementation"""
    shape = rs[:,:,0,0].shape
    nx, ny = shape[1], shape[0]
    source_number = nx * ny
    gaussian = np.zeros((source_number, source_number))
    for i in range(nx):
        for j in range(ny):
            k = kfromij(i, j, nx)
            for u in range(nx):
                for v in range(ny):
                    kl = kfromij(u, v, nx)
                    dx = i - u
                    dy = j - v
                    gaussian[k, kl] = (1 / (2 * np.pi * sigma ** 2)) * np.exp(-(dx ** 2 + dy ** 2) / (2 * sigma ** 2))
    if normalized==True: #normalizing the volume under the gaussians so that no matter the value of sigma the optimum value of lambda remains the same
        row_sums = gaussian.sum(axis=1)
        gaussian = gaussian / row_sums[:, np.newaxis]        
    return np.linalg.inv(gaussian)

def compute_cost(ps,Us_obs,G,lamb_cov_inv): #function to compute the square of the residual
    return np.sum((Us_obs-np.matmul(G,ps))**2)

def compute_grad_misfit(ps,Us_obs,G,lamb_cov_inv): #function to compute the gradient of the misfit
    return np.matmul(G.T,(np.matmul(G,ps)-Us_obs)) + np.matmul(lamb_cov_inv,ps)

def create_bounds(constraint):
    '''Function to build the bounds from the constraint vector to be used with scipy minimization approaches
    -> only used for L-BFGS because scipy doesn't have NLCG and CG doesn't accept bounds...
    :input: 1D array containing either a number (float/int) to serve as the upper and lower bound for the associated source load, or a np.nan
    if no bound is provided
    :output: list of lower and upper bounds to feed to scipy minimize function'''
    bounds = []
    for i in range(len(constraint)):
        if np.isnan(constraint[i])==False:
            bounds.append((constraint[i],constraint[i]))
        else:
            bounds.append((None,None))
    return bounds

def log_run(E,v,lamb,epsilon,sigma,constraint,it,maxit,elapsed_time,logpath):
    """ Creates a log file for the inversion """
    #not gonna put the coordinates in the file assumed to be known because we always use the same G matrix.
    if isinstance(constraint,np.ndarray)==True:
        constraint = True #donc soit None soit True
    dic = {
    'E': E,
    'v': v,
    'epsilon': epsilon,
    'lamb' : lamb, 
    'sigma' : sigma,
    'constraint' : constraint,
    'iterations' : it,
    'maxit' : maxit,
    'elapsed_time' : elapsed_time
}
    filename = f'{logpath}/meta_sig_{sigma}_lamb_{lamb}.json'    
    with open(filename, 'w') as file:
        json.dump(dic, file)


def lbfgs_direction(grad, s_hist, y_hist, rho_hist, m):
    '''Handles the memory for the LBFGS algorithm'''
    q = grad.copy()
    alpha = [0] * m #adapting the size of alpha to also work when k is smaller than m
    for i in range(len(s_hist) - 1, -1, -1):
        alpha[i] = rho_hist[i] * np.dot(s_hist[i], q)
        q -= alpha[i] * y_hist[i]
    gk = np.dot(s_hist[-1], y_hist[-1]) / np.dot(y_hist[-1], y_hist[-1])  #gamma k  used to compute H0k so that when doing linesearch alpha = 1 is often accepted as explained in Nocedal
    r = gk * q #we chose H0 = gamma k*Identity  so r is just gk * q  
    for i in range(len(s_hist)):
        beta = rho_hist[i] * np.dot(y_hist[i], r)
        r += (alpha[i] - beta) * s_hist[i]
    return r

############################################### 

def disp2load(E,v,rs,xs, ys, Us, mode=None, lamb=1, epsilon=1e-2, gamma_coeff=0.1,sigma=1,G=None,Cm1=None,
             constraint=None,maxit=1e6,verbose=0,logpath=None):
    '''Need to provide the fucntion the shape of ps expected and Us the data in the format [[x1,y1,z1],
                                                                                            [x2,y2,z2],
                                                                                            ...
                                                                                            [xn,yn,zn]]
                                                                                            
            For  mode=1  alpha is the value of lmabda, the coefficient of the regularizer                                                                        '''
    # Computing the Lame constants
    l = E * v / ((1 + v) * (1 - 2 * v))
    m = E / (2 * (1 + v))    
    
    data_number = len(Us[:,0])*len(Us[0,:])
    source_number = len(rs[:,0,0,0])*len(rs[0,:,0,0])
    
    Us_formatted = Us.reshape((data_number,1)) # trois premier les éléments de la première ligne, puis 3 suivant ce sont les 3 de la seconde ligne..
    if isinstance(G, np.ndarray)==False: #checking if G already precomputed and uses it if so : massive time gain for L-curve...
        G = build_G(rs,xs,ys,l,m)
    
    if isinstance(constraint,np.ndarray)==True: #checking if there are constraints 
        constraint = constraint.reshape(source_number) #reshaping the constraint to match that of ps = inverted parameters vector
        constraint_idx = np.where(np.isnan(constraint)==False)[0]
    
    t1 = time.time() #
    ############################### INVERSION STARTING HERE ######################
    
    if mode==None: #without regularization    
        ps = np.linalg.solve(G.T@G,G.T@Us_formatted)
        
    elif mode.lower()=='linear': #linear inversion -> non non linear approach 
        if isinstance(Cm1, np.ndarray)==False: #on apeut déjà avoir créé au préalable la matrice de covariance 
            if Cm1=='laplacian' or Cm1=='Laplacian':
                Cm1 =  build_laplacian(rs)
            elif Cm1=='Gaussian' or Cm1=='gaussian':
                Cm1 = build_gaussian_inv(rs,sigma)
        GTG = G.T@G
        GTD = G.T@Us_formatted
        ps = np.linalg.solve(GTG+lamb*Cm1.T@Cm1,GTD)
    #################################################################################################################################
    elif mode.lower()=='tv': #regularization with Total variation (TV)
        Gstar = G
        G = Gstar.T
        ps = np.zeros((source_number,1))
        b = G@Us_formatted
        T = lamb #threshold pour le  soft thresholding -> plus le threshold est faible moins on lisse!!! 
        hessian = G@Gstar
        gamma = gamma_coeff/np.max(np.abs(hessian))
        k = 0
        conv = False
        while conv==False and k<maxit:#
            ps_old = ps.copy()
            #gradient descent
            ps_bar = ps+gamma*(b-hessian@ps)
            #soft thresholding 
            for i in range(source_number):
                ps[i,0] = ps_bar[i,0]*np.max([1-(gamma*T)/np.abs(ps_bar[i,0]),0])
            res = ((ps-ps_old).T@(ps-ps_old))[0][0]
            if  res < epsilon: 
                conv = True    
            k += 1 
    #################################################################################################################################
    elif mode.lower()=='steepest': #Steepest descent 
        Us_formatted = Us.reshape((data_number,)) #reshaping data to be match the following implementations that us shape n, or m,
        if isinstance(Cm1, np.ndarray)==False: #on apeut déjà avoir créé au préalable la matrice de covariance 
            if Cm1=='laplacian' or Cm1=='Laplacian':
                Cm1 =  build_laplacian(rs)
            elif Cm1=='Gaussian' or Cm1=='gaussian':
                Cm1 = build_gaussian_inv(rs,sigma)
        lamb_cov_inv = lamb*Cm1.T@Cm1 
        alpha = 1
        conv = False
        i = 0
        ps = np.ones((source_number,))
        while conv==False and i<maxit:
            i +=1 
            delta_m = -compute_grad_misfit(ps=ps,Us_obs=Us_formatted,G=G,lamb_cov_inv=lamb_cov_inv)
            alpha_old = alpha
            alpha = linesearch(alpha=alpha_old,Us_obs=Us_formatted,G=G,ps=ps,delta_m=delta_m,lamb_cov_inv=lamb_cov_inv)
            if alpha==None:
                alpha = linesearch_armijo(alpha=alpha_old,Us_obs=Us_formatted,G=G,ps=ps,delta_m=delta_m,lamb_cov_inv=lamb_cov_inv) #mettre juste condition d'armijo si le strong wolfe n'arrive pas à converger
            if alpha<1e-10:
                alpha=1e-10
            ps += alpha*delta_m
            if isinstance(constraint,np.ndarray)==True:
                ps[constraint_idx] = constraint[constraint_idx] #on rajoute la constrainte -> par ex mettre à 0 la région souaitée 
            if np.dot(alpha*delta_m,alpha*delta_m) < epsilon:
                conv = True
    #################################################################################################################################
    elif mode.lower()=='nlcg': #mode NLCG -> faudra refaire au propre sans doute -> en appelant ces approches NLCG fast
        Us_formatted = Us.reshape((data_number,))
        ps = np.ones((source_number,))
        if isinstance(Cm1, np.ndarray)==False: #on apeut déjà avoir créé au préalable la matrice de covariance 
            if Cm1=='laplacian' or Cm1=='Laplacian':
                Cm1 =  build_laplacian(rs)
            elif Cm1=='Gaussian' or Cm1=='gaussian':
                Cm1 = build_gaussian_inv(rs,sigma)
        lamb_cov_inv = lamb*Cm1.T@Cm1
        rk = compute_grad_misfit(ps=ps,Us_obs=Us_formatted,G=G,lamb_cov_inv=lamb_cov_inv)
        pk = -rk
        conv = False
        alpha = 1
        i = 0
        while conv==False and i<maxit:
            Gpk = np.matmul(G,pk)
            #precompute the  r.T@r
            rkTrk = np.dot(rk,rk)
            alpha_old = alpha
            alpha = linesearch(alpha=alpha_old,Us_obs=Us_formatted,G=G,ps=ps,delta_m=pk,lamb_cov_inv=lamb_cov_inv)
            if alpha==None:
                alpha = linesearch_armijo(alpha=alpha_old,Us_obs=Us_formatted,G=G,ps=ps,delta_m=pk,lamb_cov_inv=lamb_cov_inv) #mettre juste condition d'armijo si le strong wolfe n'arrive pas à converger
            ps += alpha*pk
            if isinstance(constraint,np.ndarray)==True: #checking if a constraint was added
                ps[constraint_idx] = constraint[constraint_idx] #on rajoute la constrainte -> par ex mettre à 0 la région souaitée 
            rk = compute_grad_misfit(ps=ps,Us_obs=Us_formatted,G=G,lamb_cov_inv=lamb_cov_inv)
            beta = np.dot(rk,rk)/rkTrk
            pk = -rk + beta*pk
            i += 1
            if np.dot(alpha*pk,alpha*pk) < epsilon:
                conv = True
    #################################################################################################################################
    elif mode.lower()=='nlcg_fast':
        Us_formatted = Us.reshape((data_number,))
        ps = np.ones((source_number,))
        if isinstance(Cm1, np.ndarray)==False: #on apeut déjà avoir créé au préalable la matrice de covariance 
            if Cm1=='laplacian' or Cm1=='Laplacian':
                Cm1 =  build_laplacian(rs)
            elif Cm1=='Gaussian' or Cm1=='gaussian':
                Cm1 = build_gaussian_inv(rs,sigma)
        lamb_cov_inv = lamb*Cm1.T@Cm1
        result = minimize(fun=compute_cost, x0=ps, args=(Us_formatted,G,lamb_cov_inv), method='CG', tol=epsilon, jac=compute_grad_misfit,
                          options={'maxiter': maxit})
        ps = result.x
    #################################################################################################################################
    elif mode.lower()=='l-bfgs' or mode.lower()=='lbfgs': #mode L-BFGS
        Us_formatted = Us.reshape((data_number,))
        ps = np.ones((source_number,))
        if isinstance(Cm1, np.ndarray)==False: #on apeut déjà avoir créé au préalable la matrice de covariance 
            if Cm1=='laplacian' or Cm1=='Laplacian':
                Cm1 =  build_laplacian(rs)
            elif Cm1=='Gaussian' or Cm1=='gaussian':
                Cm1 = build_gaussian_inv(rs,sigma)
        lamb_cov_inv = lamb*Cm1.T@Cm1
        i = 0
        m = 5 #number of s and y pairs to keep to act as the hessian estimate, 5 is supposed to be enough based on Nocedal
        conv = False
        s_hist = []
        y_hist = []
        rho_hist = []
        q = compute_grad_misfit(ps=ps,Us_obs=Us_formatted,G=G,lamb_cov_inv=lamb_cov_inv)
        while conv==False and i<maxit:
            if i == 0:
                pk = -q
            else:
                pk = -lbfgs_direction(q, s_hist, y_hist, rho_hist, m)
            alpha = linesearch(alpha=1,Us_obs=Us_formatted,G=G,ps=ps,delta_m=pk,lamb_cov_inv=lamb_cov_inv) #start with alpha=1 because we use the scaler for pk
            if alpha==None:
                alpha = linesearch_armijo(alpha=1,Us_obs=Us_formatted,G=G,ps=ps,delta_m=pk,lamb_cov_inv=lamb_cov_inv) #mettre juste condition d'armijo si le strong wolfe n'arrive pas à converger
            if alpha<1e-10:
                alpha=1e-10 #if still managed to fail to estimate alpha then we use this ad hoc value 
            ps_old = ps.copy()
            ps += alpha*pk
            if isinstance(constraint,np.ndarray)==True:
                ps[constraint_idx] = constraint[constraint_idx] #on rajoute la constrainte -> par ex mettre à 0 la région souaitée 
            sk = ps-ps_old #vecteur de shape source_number,
            q_old = q.copy()
            q = compute_grad_misfit(ps=ps,Us_obs=Us_formatted,G=G,lamb_cov_inv=lamb_cov_inv)
            yk = q-q_old #vecteur de shape source_number,
            rho_k = 1 / np.dot(yk, sk)
            if len(s_hist) == m:
                s_hist.pop(0)
                y_hist.pop(0)
                rho_hist.pop(0)
            s_hist.append(sk)
            y_hist.append(yk)
            rho_hist.append(rho_k)
            i += 1
            if np.dot(alpha*pk,alpha*pk) < epsilon:
                conv = True
    #################################################################################################################################
    elif mode.lower()=='l-bfgs_fast' or mode.lower()=='lbfgs_fast': #mode L-BFGS fast avec scipy alsos uses Wolfe conditions to asses the value of alpha
        Us_formatted = Us.reshape((data_number,))
        ps = np.ones((source_number,))
        if isinstance(Cm1, np.ndarray)==False: #on apeut déjà avoir créé au préalable la matrice de covariance 
            if Cm1=='laplacian' or Cm1=='Laplacian':
                Cm1 =  build_laplacian(rs)
            elif Cm1=='Gaussian' or Cm1=='gaussian':
                Cm1 = build_gaussian_inv(rs,sigma)
        lamb_cov_inv = lamb*Cm1.T@Cm1
        if isinstance(constraint,np.ndarray)==True: #checking if a constraint was added
            ps[constraint_idx] = constraint[constraint_idx] #we make the first guess start with the correct value
            bounds = create_bounds(constraint)#we create bounds equal to our constraints for the specific ps values that need it 
            result = minimize(fun=compute_cost, x0=ps, args=(Us_formatted,G,lamb_cov_inv), method='L-BFGS-B', tol=epsilon, jac=compute_grad_misfit,
                         bounds=bounds,options={'maxiter': maxit})
        else:
            result = minimize(fun=compute_cost, x0=ps, args=(Us_formatted,G,lamb_cov_inv), method='L-BFGS-B', tol=epsilon, jac=compute_grad_misfit)
        ps = result.x
    
    if logpath!=None:
        elapsed_time = time.time()-t1
        if mode.lower()=='linear' or mode==None:
            i = None
        elif mode.lower()=='lbfgs_fast' or mode.lower()=='nlcg_fast': 
            i = result.nit
        log_run(E,v,lamb,epsilon,sigma,constraint,i,maxit,elapsed_time,logpath)
    
    ps = ps.reshape(rs[:,:,0,0].shape)
    return ps
```

### 2D verification with the results from d'urso and Marmo

In this section we try to replicate the result obtained in D'Urso and Marmo, 2013. For that purpose we use the same complex source shape corresponding to the The Great Salt Lake in Utah with the same parameters, that is to say E = 75GPa, $\nu$ = 0.25 and a water, $\rho_{water}$ = 1150 kg/m$^3$ and a constant water level of 1 m.

```python
#we can model the displacements with 2 different approaches; either we can consider a signal source as was previously done for 1D approach or
#we can create a myriad of small simple sources, which is the approach required for inversion.
p = 1150*9.81*1 #rho g z
####################### approach with a single source ##########################################################
s = pd.read_csv(f'true_source.csv') #file containing the coordinates of the contour of the lake
lat_coordinates = s['lat'].values
lon_coordinates = s['lon'].values
x_coordinates, y_coordinates = utils.latlon2xy(lat_coordinates, lon_coordinates, zone=11) #converting to cartesian coordinates
#We now have all the coordinates 
r = np.zeros((len(x_coordinates),2))
r[:,0] = x_coordinates
r[:,1] = y_coordinates

##################### approach with multiple small sources ##################################################
def is_point_inside_polygon(polygon, x, y):
    """Function that allows to determine if a point is inside a polygon : useful to map the pressure distribution front contours """
    n = len(polygon)
    inside = False
    p1x, p1y = polygon[0]
    for i in range(n + 1):
        p2x, p2y = polygon[i % n]
        if y > min(p1y, p2y):
            if y <= max(p1y, p2y):
                if x <= max(p1x, p2x):
                    if p1y != p2y:
                        xints = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x
                    if p1x == p2x or x <= xints:
                        inside = not inside
        p1x, p1y = p2x, p2y
    return inside

ps = np.zeros((100,100)) #gonna create a mesh of 100*100 sources
#need to match the extent from the figure of D'Urso and Marmo to easily compare our results
xstart, ystart = utils.latlon2xy(40.5,-113.5,zone=11)
xend, yend = utils.latlon2xy(42,-111.5,zone=11)
dx = (xend-xstart)/100
dy = (yend-ystart)/100
rs = utils.create_source_mesh(xstart,xend,xstart,yend,ps) #sources should have part out of the domain in which we compute 
xr = np.linspace(xstart+dx/2,xend-dx/2,100)
yr = np.linspace(ystart+dy/2,yend-dy/2,100)
# xr = np.linspace((rs[0,0,1,0]+rs[0,0,0,0])/2, (rs[0,-1,0,0]+rs[0,-1,1,0])/2 ,len(ps[0,:]))
# yr = np.linspace((rs[0,0,3,1]+rs[0,0,0,1])/2, (rs[-1,0,3,1]+rs[-1,0,0,1])/2, len(ps[:,0])) #les 

for j in range(len(ps[:,0])):
    for i in range(len(ps[0,:])):
        result = is_point_inside_polygon(r, xr[i], yr[j]) #we loop over the sources and tell if they are inside the polygon r
        if result == True:
            ps[j,i] = p  #if inside the lake then we use the load from 1m of water of rho = 1150 kg/m^3

########### plotting the resulting load distributions #######
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].plot(r[:,0],r[:,1],color='black')
ax[0].fill(r[:,0],r[:,1], color='yellow')
ax[0].patch.set_facecolor('purple')
ax[0].set_xlim(xr[0],xr[-1])
ax[0].set_ylim(yr[0],yr[-1])
im = ax[1].pcolormesh(xr,yr,ps)
# cbar = plt.colorbar(im)
```

```python
levels = np.arange(0,8.5,0.5) #levels for the colorbar 

fig, ax = plt.subplots(2,3,figsize=(14,6))
ax[0,0].set_title('1 complex source vertical')
ax[0,0].set_xlabel('Longitude (°)')
ax[0,0].set_ylabel('Latitude (°)')
im = ax[0,0].contourf(lons,lats,Us_unique[:,:,2]*1000,levels=levels, cmap='jet') #plotting the vetical displacement in the 1st row
cbar= plt.colorbar(im)
cbar.set_label(r'd$_z$ (mm)')
ax[0,0].plot(lon_coordinates,lat_coordinates, color='black',linewidth=2) #plotting the contour of the lake

ax[0,1].set_title('Multiple simple sources vertical')
ax[0,1].set_xlabel('Longitude (°)')
ax[0,1].set_ylabel('Latitude (°)')
im = ax[0,1].contourf(lons,lats,Us_multiple[:,:,2].T*1000,levels=levels, cmap='jet') #plotting the vetical displacement in the 1st row
# ax[0,1].locator_params(nbins=6,axis='x')
cbar= plt.colorbar(im)
cbar.set_label(r'd$_z$ (mm)')
ax[0,1].plot(lon_coordinates,lat_coordinates, color='black',linewidth=2) #plotting the contour of the lake

ax[0,2].axis('off')
ax[0,2].imshow(mpimg.imread('vertical.png')) #using the plot from d'urso and marmo 
ax[0,2].set_title("Vertical from D'Urso and Marmo")

###########################################################################
levels = np.arange(0,1.3,0.1)

ax[1,0].set_title('1 comple source horizontal')
ax[1,0].set_xlabel('Longitude (°)')
ax[1,0].set_ylabel('Latitude (°)')
im = ax[1,0].contourf(lons,lats,np.sqrt( (Us_unique[:,:,1]*1000)**2 + (Us_unique[:,:,0]*1000)**2 ),levels=levels, cmap='jet')
# ax[1,0].locator_params(nbins=6,axis='x')
cbar=  plt.colorbar(im)
cbar.set_label(r'$|\delta|$ (mm)')
ax[1,0].plot(lon_coordinates,lat_coordinates, color='black',linewidth=2)
x_subset = Us_unique[::reduction_factor, ::reduction_factor,0]
y_subset = Us_unique[::reduction_factor, ::reduction_factor,1]
ax[1,0].quiver(XX[::reduction_factor, ::reduction_factor],YY[::reduction_factor, ::reduction_factor],x_subset,y_subset, scale=0.02)

ax[1,1].set_title('Multiple simple sources horizontal')
ax[1,1].set_xlabel('Longitude (°)')
ax[1,1].set_ylabel('Latitude (°)')
im = ax[1,1].contourf(lons,lats,np.sqrt( (Us_multiple[:,:,1]*1000)**2 + (Us_multiple[:,:,0]*1000)**2 ).T,levels=levels, cmap='jet')
# ax[1,1].locator_params(nbins=6,axis='x')
cbar=  plt.colorbar(im)
cbar.set_label(r'$|\delta|$ (mm)')
ax[1,1].plot(lon_coordinates,lat_coordinates, color='black',linewidth=2)
x_subset = Us_multiple[::reduction_factor, ::reduction_factor,0]
y_subset = Us_multiple[::reduction_factor, ::reduction_factor,1]
ax[1,1].quiver(XX[::reduction_factor, ::reduction_factor],YY[::reduction_factor, ::reduction_factor],x_subset,y_subset, scale=0.02)

ax[1,2].axis('off')
ax[1,2].imshow(mpimg.imread('horizontal.png')) #using the plot from d'urso and marmo 
ax[1,2].set_title("horizontal from D'Urso and Marmo")

fig.subplots_adjust(wspace=0.4, hspace=0.5)
```

This example illustrates the fact that the contribution of the multiple small sources to the overall displacmenent can indeed sum up linearly and are equivalent to the displacement from a single complex source. ### Simple load distribution with low noise 

```python
#ici on fait déjà avec small noise et 3 approches et comparaison avec le ground truth
# en utilisant inversion linéaire #pas de non linéaire avec autant de pixels car ça seait long
#faut que l'on fasse L-curve et tout  et tester différents std pour pouvoir trouver le meilleur 
seed_value = 42
np.random.seed(seed_value)
xs = np.random.randint(-2000, high=2000, size=50) #les x des stations
ys =  np.random.randint(-2000, high=2000, size=50) #les y des stations 
z = 0 #va mettre toutes les stations à 0 m
Us = np.zeros((len(xs), 3)) #on va faire seulement une ligne de displaements dans ce cas car on a pas des mesures régulières !
#defining the region where pressure is applied
r = np.array([[-500., -1000.], [500., -1000.], [500., 1000.], [-500., 1000.]])
p = 9.82*10**5
E = 0.6e11
v = 0.25
###### pour pouvoir faire la version avec plusieurs sources au lieu d'une seule source
ps = np.zeros((8,8)) 
ps[0:2,0:2] = 9.81*10**3
ps[-2:,-2:] = 9.81*10**3
rs = utils.create_source_mesh(-2000,2000,-2000,2000,ps)

for vy_idx in range(len(ps[:,0])): #boucle sur les sources et on va sommet les réusltats sur les vectices pour faire le délacement gloabel 
    for vx_idx in range(len(ps[0,:])): 
        p = ps[vy_idx,vx_idx]
        if p!=0: #si p =0 alors pas besoin de calculer l'influence de cette source car nulle 
            r = rs[vy_idx,vx_idx]
            for i, x in enumerate(xs):
                xyz = [xs[i], ys[i], z]
                U = load2disp.load2disp( xyz, r, p, E, v)
                Us[ i, :] += U.reshape(3)
                
#We now add noise to the displacements
seed_value = 42 
np.random.seed(seed_value)
Us_noisy = Us.copy()
Us_noisy[:,0:2] += np.random.normal( loc=0.0, scale=0.05*np.max(np.abs(Us[:,0:2])), size=Us[:,0:2].shape ) #pourrait faire noise plus ou moins fort sur les composantes -> par ex 
Us_noisy[:,2] += np.random.normal( loc=0.0, scale=0.05*np.max(np.abs(Us[:,2])), size=Us[:,2].shape ) #pourrait faire noise plus ou moins fort sur les composantes -> par ex 

#on pourrait mettre un niveau de bruit plus fort sur composante verticale  et même lui donner une valeur de bruit qui est logique  avec ce qui s'observe ...
vmax = np.max([np.abs(Us_noisy[:,2]),np.abs(Us[:,2])])*1000
arrow_coeff = 100000

fig, ax = plt.subplots(1, 3,figsize=(15,3))
im = ax[0].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0].scatter(xs,ys,marker='^',linewidths=1, edgecolors='black', color='white')
ax[0].set_xlabel('x (m)')
ax[0].set_ylabel('y (m)')
ax[0].locator_params(nbins=6,axis='x')

ax[1].quiver(xs, ys, Us[:,0]*arrow_coeff, Us[:,1]*arrow_coeff, color='black')
im = ax[1].scatter(xs,ys,c=Us[:,2]*1000,label='z displacement', vmax=vmax, vmin=-vmax, cmap='seismic') 
cbar = plt.colorbar(im)
cbar.set_label(r'd$_z$ (mm) ')
ax[1].set_title('displacement without noise')
ax[1].set_ylabel('y (m)')
ax[1].set_xlabel('x (x)')
ax[1].locator_params(nbins=6,axis='x')
ax[1].set_xlim(-2000,2000)
ax[1].set_ylim(-2000,2000)

ax[2].quiver(xs, ys, Us_noisy[:,0]*arrow_coeff, Us_noisy[:,1]*arrow_coeff, color='black')
im = ax[2].scatter(xs,ys,c=Us_noisy[:,2]*1000,label='z displacement', vmax=vmax, vmin = -vmax, cmap='seismic') 
cbar = plt.colorbar(im)
cbar.set_label(r'd$_z$ (mm) ')
ax[2].set_title('displacement with noise')
ax[2].set_ylabel('y (m)')
ax[2].set_xlabel('x (x)')
ax[2].locator_params(nbins=6,axis='x')
ax[2].set_xlim(-2000,2000)
ax[2].set_ylim(-2000,2000)

fig.subplots_adjust(wspace=0.5)
```

```python
#now we perform the inversion with the different approaches and compare to the actual load distribution 
#looping over various values of lambda to find the best value of lambda for each implementation via the L curve 
#performing a L curve to determine the best value for alpha 
data_number = len(Us[:,0])*3 #number of receivers *3 
source_number = ps.size
l = E * v / ((1 + v) * (1 - 2 * v))
m = E / (2 * (1 + v))    
disp_obs = Us_noisy.reshape((data_number,1))
G = disp2load.build_G(rs,xs,ys,l,m)
laplacian = disp2load.build_laplacian(rs)
TV = np.sqrt(np.abs(laplacian))


lambs = np.logspace(-21,0,80)
sigmas = np.logspace(-6,3,50) #looping over sigmas to also find the 
gaussians = np.zeros((source_number,source_number,len(sigmas)))
for i,sigma in enumerate(sigmas):
    gaussians[:,:,i] = disp2load.build_gaussian_inv(rs,sigma=sigma,normalized=True)
norm_laplacian_ps = np.zeros(len(lambs)) #calcul de la norme des solutions
norm_residuals_laplacian = np.zeros(len(lambs)) #calcul de la norme des résidus
norm_diff_ps_laplacian = np.zeros(len(lambs)) #calcul de la norme des résidus
norm_TV_ps = np.zeros(len(lambs))
norm_residuals_TV = np.zeros(len(lambs))
norm_diff_ps_TV = np.zeros(len(lambs)) 
norm_gaussian_ps = np.zeros((len(sigmas),len(lambs)))
norm_residuals_gaussian = np.zeros((len(sigmas),len(lambs)))
norm_diff_ps_gaussian = np.zeros((len(sigmas),len(lambs))) 


for i in trange(len(lambs)):
    lamb = lambs[i]
    ##### LAPLACIAN COMPUTAITONS #######
    ps_laplacian = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=lamb,Cm1=laplacian,G=G) #precomputed G and lalaplcian to save time
    ps_laplacian_reshaped = ps_laplacian.reshape((source_number,1))
    norm_laplacian_ps[i] = np.sqrt(np.sum((laplacian@ps_laplacian_reshaped)**2)) 
    disp_cal = G@ps_laplacian_reshaped
    norm_residuals_laplacian[i] = np.sqrt(np.sum((disp_obs-disp_cal)**2))
    norm_diff_ps_laplacian[i] =  np.sqrt(np.sum((ps-ps_laplacian)**2))    
    ###### TV COMPUTAITONS #######
    ps_TV = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='TV', lamb=lamb, epsilon=1e-0, gamma_coeff=0.1,G=G) #for TV need to use very low values of lambda
    ps_TV_reshaped = ps_TV.reshape((source_number,1))
    norm_TV_ps[i] = np.sqrt(np.sum((TV@ps_TV_reshaped)**2)) 
    disp_cal = G@ps_TV_reshaped
    norm_residuals_TV[i] = np.sqrt(np.sum((disp_obs-disp_cal)**2))
    norm_diff_ps_TV[i] =  np.sqrt(np.sum((ps-ps_TV)**2))   
    ##### GAUSSIAN COMPUTATIONS ######    
    for j,sigma in enumerate(sigmas):
        gaussian = gaussians[:,:,j]
        ps_gaussian = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=lamb,G=G,Cm1=gaussian)
        ps_gaussian_reshaped = ps_gaussian.reshape((source_number,1))
        norm_gaussian_ps[j,i] = np.sqrt(np.sum((gaussian@ps_gaussian_reshaped)**2)) 
        disp_cal = G@ps_gaussian_reshaped
        norm_residuals_gaussian[j,i] = np.sqrt(np.sum((disp_obs-disp_cal)**2))
        norm_diff_ps_gaussian[j,i] =  np.sqrt(np.sum((ps-ps_gaussian)**2))   
    

#### last computation ###
### laplacian finding the best value of lambda and the one predicted by the ;curve 
idx_lambda_optimum_laplacian = np.argmin(norm_diff_ps_laplacian)
def second_deriv_Lcurve(norm_regu,misfit,lambda_list,lambda_range):
    """Computes the second derivative of the L curve -> allows to detect the elbow of the L-curve"""
    sorted_indices = np.argsort(norm_regu)
    x_sorted = norm_regu[sorted_indices]
    y_sorted = misfit[sorted_indices]
    lambda_list_sorted = lambda_list[sorted_indices]
    idx2keep = np.where((lambda_list_sorted>=lambda_range[0])&(lambda_list_sorted<=lambda_range[1]))[0]
    x_sorted_cropped = x_sorted[idx2keep]
    y_sorted_cropped = y_sorted[idx2keep]
    lambda_list_cropped = lambda_list_sorted[idx2keep]
    dy_dx = np.gradient(y_sorted_cropped, x_sorted_cropped)
    d2y_dx2 = np.gradient(dy_dx, x_sorted_cropped)
    return x_sorted_cropped, d2y_dx2, lambda_list_cropped

x_laplacian, d2_laplacian, lambda_list_laplacian = second_deriv_Lcurve(norm_laplacian_ps,norm_residuals_laplacian,lambs,[1e-9,1e-6]) #comp
idx_lambda_Lcurve_laplacian = np.argmax(d2_laplacian) #we can see that we don't find quite the same value of lambda between the truth and the Lcurve, but its close enough with the noise level we have
print(fr'relative error on $\lambda$ with Laplacian approach : {100*np.abs(lambs[idx_lambda_optimum_laplacian]-lambda_list_laplacian[idx_lambda_Lcurve_laplacian])/np.min([lambs[idx_lambda_optimum_laplacian],lambda_list_laplacian[idx_lambda_Lcurve_laplacian]])} (%)')

### TV
idx_lambda_optimum_TV = np.argmin(norm_diff_ps_TV)
x_TV, d2_TV, lambda_list_TV = second_deriv_Lcurve(norm_TV_ps,norm_residuals_TV,lambs,[5e-15,5e-13]) #comp
idx_lambda_Lcurve_TV = np.argmax(d2_TV) #we can see that we don't find quite the same value of lambda between the truth and the Lcurve, but its close enough with the noise level we have
print(fr'relative error on $\lambda$ with TV approach : {100*np.abs(lambs[idx_lambda_optimum_TV]-lambda_list_TV[idx_lambda_Lcurve_TV])/np.min([lambs[idx_lambda_optimum_TV],lambda_list_TV[idx_lambda_Lcurve_TV]])} (%)')

### Gaussian
idx_sigma_correct = np.argmin(np.abs(sigmas-5e-1)) #we assume that 5e-1 is the correct value for sigma, we will have to perform an analysis afterwards to verify that 
#what value of sigma should be used to have the best estimate from the L-curve
idx_lambda_optimum_gaussian = np.argmin(norm_diff_ps_gaussian[idx_sigma_correct,:]) #finding the index of lambda for the chosen standard deviation
x_gaussian, d2_gaussian, lambda_list_gaussian = second_deriv_Lcurve(norm_gaussian_ps[idx_sigma_correct,:],
                                                                    norm_residuals_gaussian[idx_sigma_correct,:],lambs,[1e-19,1e-16]) #comp
idx_lambda_Lcurve_gaussian = np.argmax(d2_gaussian) #we can see that we don't find quite the same value of lambda between the truth and the Lcurve, but its close enough with the noise level we have

##### plotting the result

fig, ax = plt.subplots(3, 3,figsize=(14,9))
######## LAPLACIAN #### 
ax[0,0].set_title('Laplacian approach')
ax[0,0].loglog(lambs, norm_diff_ps_laplacian,linewidth=0, marker='.')
ax[0,0].loglog(lambs[idx_lambda_optimum_laplacian], norm_diff_ps_laplacian[idx_lambda_optimum_laplacian],linewidth=0, marker='.',color='red')
ax[0,0].set_xlabel(r'$\lambda$')
ax[0,0].set_ylabel(r'$|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$') #ps is the true load distribution 
im = ax[1,0].scatter(norm_laplacian_ps, norm_residuals_laplacian,c=lambs, cmap='turbo', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[1,0].set_xlabel(r'$|| \nabla \mathbf{p} ||^2_2$')
ax[1,0].set_ylabel(r'$|| \mathbf{G p} - \mathbf{d} ||_2^2$')
im = ax[2,0].scatter(x_laplacian, d2_laplacian,c=lambda_list_laplacian, cmap='viridis', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[2,0].set_xlabel(r'$|| \nabla \mathbf{p} ||^2_2$')
ax[2,0].set_ylabel(r'$ second deriv || \mathbf{G p} - \mathbf{d} ||_2^2$')
######## TV ###################### -> no real possiblity to go for the L curve 
ax[0,1].set_title('TV approach')
ax[0,1].loglog(lambs, norm_diff_ps_TV,linewidth=0, marker='.')
ax[0,1].loglog(lambs[idx_lambda_optimum_TV], norm_diff_ps_TV[idx_lambda_optimum_TV],linewidth=0, marker='.',color='red')
ax[0,1].set_xlabel(r'$\lambda$')
ax[0,1].set_ylabel(r'$|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$') #ps is the true load distribution 
im = ax[1,1].scatter(norm_TV_ps, norm_residuals_TV,c=lambs, cmap='turbo', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[1,1].set_xlabel(r'$|| \nabla \mathbf{p} ||_1$')
ax[1,1].set_ylabel(r'$|| \mathbf{G p} - \mathbf{d} ||_2^2$')
im = ax[2,1].scatter(x_TV, d2_TV,c=lambda_list_TV, cmap='viridis', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[2,1].set_xlabel(r'$|| \nabla \mathbf{p} ||_1$')
ax[2,1].set_ylabel(r'$ second deriv || \mathbf{G p} - \mathbf{d} ||_2^2$')

######### GAUSSIAN ###########
ax[0,2].set_title('Gaussian approach')
im = ax[0,2].pcolormesh(lambs,sigmas,norm_diff_ps_gaussian)
cbar = plt.colorbar(im)
cbar.set_label(r'$|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$')
ax[0,2].set_xlabel(r'$\lambda$')
ax[0,2].set_ylabel(r'$\sigma$') #ps is the true load distribution 
ax[0,2].set_xscale('log')
ax[0,2].set_yscale('log')
im = ax[1,2].scatter(norm_gaussian_ps[idx_sigma_correct,:], norm_residuals_gaussian[idx_sigma_correct,:],c=lambs, cmap='turbo', norm=matplotlib.colors.LogNorm(),label=fr'$\sigma$={sigmas[idx_sigma_correct]}')
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[1,2].set_xlabel(r'$|| Gauss \mathbf{p} ||^2_2$')
ax[1,2].set_ylabel(r'$|| \mathbf{G p} - \mathbf{d} ||_2^2$')
ax[1,2].legend()
im = ax[2,2].scatter(x_gaussian, d2_gaussian,c=lambda_list_gaussian, cmap='viridis', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[2,2].set_xlabel(r'$|| Gauss \mathbf{p} ||^2_2$')
ax[2,2].set_ylabel(r'$ second deriv || \mathbf{G p} - \mathbf{d} ||_2^2$')

fig.subplots_adjust(wspace=0.5, hspace=0.3)
```

With the Laplacian and TV approaches we find very similar results; the lowest error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ has the same value (about 2e8 -> use seed to have constant values) and the L-curve predicts as effectively the best value of $\lambda$ with relative error with the optimum $\lambda$ close to 0%. 

On the other hand, one can see that for the Normalized Gaussian approach, when the standard deviation is too large, the error becomes very large, because the gaussian is too wide, however one can also notice that the error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ actually decreases significantly for a value of sigma close to the stability limit (appears to be easier to see for large noise level), and the optimum value of lambda actually changes for this specific value of lambda. Also, the lowest error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ reach is higher than with either Laplacian or TV, which shows that in this case, this approach cannot perform as well as the 2 others. Based on theses observations it seems that we need to investigate the behavior of the inversion for different pairs of $\sigma$ and $\lambda$.

```python
#We plot the load disitrubtion for different pairs : unstable combination because of too large sigma, unstable because of too large lambda, stable with a normal
#value of lambda and stable with the absolute best pair of sigma and lambda.

ps_large_sigma = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=1e-17,
                                  G=G,Cm1='gaussian',sigma=1e1)

ps_large_lambda = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=1e-12,
                                  G=G,Cm1='gaussian',sigma=1e-1)

ps_optimum_pair = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=lambs[idx_lambda_optimum_gaussian],
                                  G=G,Cm1='gaussian',sigma=sigmas[idx_sigma_correct])

idx_selec = np.argmin(np.abs(1e-1-sigmas))
ps_good_pair = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=lambs[np.argmin(norm_diff_ps_gaussian[idx_selec,:])],
                                  G=G,Cm1='gaussian',sigma=sigmas[idx_selec])

fig, ax = plt.subplots(1,4,figsize=(14,2))
#suing the same scale for the different schemes 
vmin = np.min([ps_large_lambda,ps_large_sigma,ps_optimum_pair,ps_good_pair])
vmax = np.max([ps_large_lambda,ps_large_sigma,ps_optimum_pair,ps_good_pair])

im = ax[0].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_optimum_pair, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0].set_xlabel('x (m)')
ax[0].set_ylabel('y (m)')
ax[0].set_title(fr'Optimum pair : $\lambda$={"{:.1e}".format(lambs[idx_lambda_optimum_gaussian])}, $\sigma$={"{:.1e}".format(sigmas[idx_sigma_correct])}')

im = ax[1].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_good_pair, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[1].set_xlabel('x (m)')
ax[1].set_ylabel('y (m)')
ax[1].set_title(fr'Good pair : $\lambda$={"{:.1e}".format(lambs[np.argmin(norm_diff_ps_gaussian[idx_selec,:])])}, $\sigma$={"{:.1e}".format(sigmas[idx_selec])}')

im = ax[2].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_large_sigma, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[2].set_xlabel('x (m)')
ax[2].set_ylabel('y (m)')
ax[2].set_title(fr'Too large $\sigma$ : $\lambda$=1e-17, $\sigma$=1e1')

im = ax[3].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_large_lambda, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[3].set_xlabel('x (m)')
ax[3].set_ylabel('y (m)')
ax[3].set_title(fr'Too large $\lambda$ : $\lambda$=1e-12, $\sigma$=1e-1')

fig.subplots_adjust(wspace=1, hspace=0.3)
```

As one can see, when one choses a value of $\sigma$ that is low enough that the inversion is stable, the result of the inversion even with the optimum value of $\lambda$ is not great, considering the low level of noise in the data. When choosing a value of $\sigma$ sigma too large, the inversion becomes unstable, leading to a a characteristic null load distribution. The same happens for a value of $\lambda$ too large. When choosing the optimum value of $\sigma$, and the optimum value of $\lambda$ associated, one finds the inverted load distribution quite good. 
  
Therefore, it seems to be more interesting to use the largest value of $\sigma$ that still leads to a stable inversion, as it appears to lead to the lowest error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$. Based on the observed behaviour, it seems relatively easy to find the best $\sigma$,since when it is too large, the inverted load distribution is null (obvious unstability). In addition, the optimum value of lambda remains constant for small values of $\sigma$, therefore one may estimate the best $\sigma$ based on the inverted load distribution and on the evolution of the predicted $\lambda$ as a function of $\sigma$. 

For that purpose, we must investigate the ability of the L-curve to correcctly estimate the value of $\lambda$ for large values of $\sigma$, to make sure that we will be able to observe the constant $\lambda$ for small $\sigma$, and the fact that $\lambda$ varies once the best value of $\sigma$ is reached.

```python
#we verify that the standard deviation we chose is the one for which the L-curve is the most effective to predict the value of lambda. For that we simply have to compute the 
#the norm between the ps inverted with the optimum lambda and the one inverted with the value of lambda foudnd with the Lcurve.
lambdas_true = np.zeros(len(sigmas))
lambdas_Lcurve = np.zeros(len(sigmas))
misfit4optimum = np.zeros(len(sigmas))

for i, sigma in enumerate(sigmas):
    lambdas_true[i] = lambs[np.argmin(norm_diff_ps_gaussian[i,:])]
    misfit4optimum[i] = np.min(norm_diff_ps_gaussian[i,:]) #also keeping that to be able to tell if the relative error of lambda is relevant 
    x, d2, lambda_list = second_deriv_Lcurve(norm_gaussian_ps[i,:],
                                            norm_residuals_gaussian[i,:],lambs,[1e-19,1e-16])
    lambdas_Lcurve[i] = lambda_list[np.argmax(d2)]  
    
relative_error = 100*np.abs(lambdas_Lcurve-lambdas_true)/np.min([lambdas_true,lambdas_Lcurve],axis=0)

print(r'The lowest error between the true and inverted load distributions is obtained for $\sigma$ = ',sigmas[np.argmin(misfit4optimum)])
print(fr'For sigma = {sigmas[np.argmin(misfit4optimum)]}, the relative error of lambda is {relative_error[np.argmin(misfit4optimum)]} %')

#adding figure to keep track of the estimated  values of lmabda and vs the optimum one 

fig, ax = plt.subplots(1,3, figsize=(14,4))

ax[0].scatter(sigmas,lambdas_true, label='Optimum')
ax[0].scatter(sigmas,lambdas_Lcurve, label='Lcurve')
ax[0].set_title(r'$\lambda$ vs $\sigma$')
ax[0].set_xlabel(r'$\sigma$')
ax[0].set_xscale('log')
ax[0].set_ylabel(r'$\lambda$')
ax[0].set_yscale('log')
ax[0].legend()

ax[1].scatter(sigmas,relative_error) #plotting the relative error between the value of lambda found with l-curve and the optimum one 
ax[1].set_ylabel(r'relative error $|\lambda_{L-curve}-\lambda_{true}|/ \lambda_{true}$ (%)')
ax[1].set_xlabel(r'$\sigma$')
ax[1].set_xscale('log')
ax[1].set_title(r'Relative error for $\lambda$')
ax[1].set_yscale('log')

ax[2].scatter(sigmas,misfit4optimum)
ax[2].set_xlabel(r'$\sigma$')
ax[2].set_ylabel(r'$|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$')
ax[2].set_xscale('log')
ax[2].set_title('Misfit for optimum lambda')
ax[2].set_yscale('log')
fig.subplots_adjust(wspace=0.3)
```

It seems that when the value of $\sigma$ is small, its value of does not actually impact the relative error between the value of $\lambda$ estimated the L-curve and the optimum one. When sigma becomes large, one can see that the L-curve starts overestimating the value of $\lambda$, where the larger $\sigma$, the more we overestimate $\lambda$ overall. 

In this example, the lowest error is obtained for a standard deviation of 0.49, and for this $\sigma$, the value of $\lambda$ predicted by the L-curve is still identical to the ones for lower $\sigma$ (the next value predicted for slightly larger $\sigma$ is different than the previous ones however), and the error between the predicted value of $\lambda$ and the optimum one is null, which shows that the L-curve works perfectly is this example. 

Based on these observations, it seems that that the L-curve predicts very well the value of $\lambda$, and that we can identify the optimal value of $\sigma$ as the value at which the $\lambda$ predicted by the L-curve diverges from what is derived for low values of $\sigma$.

In the following cell we represent the load distribution inverted using the Laplacian, TV and Gaussian regularization approaches with the value of $\lambda$ derived from the the L-curve.

```python
#in this cell we now display the result of the inversion using the value of lambda predicted by the Lcurve for the different approaches, result should be nice since
#there is little noise on the displacements 
ps_laplacian = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=lambda_list_laplacian[idx_lambda_Lcurve_laplacian],Cm1=laplacian,G=G)
ps_TV = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='TV', lamb=lambda_list_TV[idx_lambda_Lcurve_TV], epsilon=1e-8, gamma_coeff=0.1,G=G) #pour le moment use solution optimale 
ps_gaussian = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=lambda_list_gaussian[idx_lambda_Lcurve_gaussian],
                                  G=G,Cm1='gaussian',sigma=sigmas[idx_sigma_correct])

fig, ax = plt.subplots(1,4,figsize=(15,3))

#suing the same scale for the different schemes 
vmin, vmax = np.min([ps,ps_gaussian,ps_TV,ps_laplacian]), np.max([ps,ps_gaussian,ps_TV,ps_laplacian])

im = ax[0].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0].set_xlabel('x (m)')
ax[0].set_ylabel('y (m)')
ax[0].set_title('Orignal')

im = ax[3].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_laplacian, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[3].set_xlabel('x (m)')
ax[3].set_ylabel('y (m)')
ax[3].set_title('Laplacian')

im = ax[1].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_TV, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[1].set_xlabel('x (m)')
ax[1].set_ylabel('y (m)')
ax[1].set_title('TV')

im = ax[2].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_gaussian, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[2].set_xlabel('x (m)')
ax[2].set_ylabel('y (m)')
ax[2].set_title('Gaussian')

fig.subplots_adjust(wspace=0.7, hspace=0.3)
```

We can see that all approaches reproduce fairly well the original load distribution. One can note that while Laplacian finds the most logical load distribution at the sources (pixels of the sources have more uniform values), total variation better constrains spacially the sources. Meanwhile, the load distribution inverted with the Gaussian approach is farily similar to the one derived from the Laplacian, but with wose estimate of the load (consistant with the higher error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$). 

### Simple load distribution with high noise

We will now investigate the behavior of these approaches with a larger noise level, to see if they behave similarly to what was seen previously.

```python
#We now add noise to the displacements
seed_value = 42
np.random.seed(seed_value)
Us_noisy = Us.copy()
Us_noisy[:,0:2] += np.random.normal( loc=0.0, scale=0.2*np.max(np.abs(Us[:,0:2])), size=Us[:,0:2].shape ) #pourrait faire noise plus ou moins fort sur les composantes -> par ex 
Us_noisy[:,2] += np.random.normal( loc=0.0, scale=0.2*np.max(np.abs(Us[:,2])), size=Us[:,2].shape ) #pourrait faire noise plus ou moins fort sur les composantes -> par ex 

#on pourrait mettre un niveau de bruit plus fort sur composante verticale  et même lui donner une valeur de bruit qui est logique  avec ce qui s'observe ...
vmax = np.max([np.abs(Us_noisy[:,2]),np.abs(Us[:,2])])*1000
arrow_coeff = 100000

fig, ax = plt.subplots(1, 3,figsize=(15,3))
im = ax[0].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0].scatter(xs,ys,marker='^',linewidths=1, edgecolors='black', color='white')
ax[0].set_xlabel('x (m)')
ax[0].set_ylabel('y (m)')
ax[0].locator_params(nbins=6,axis='x')

ax[1].quiver(xs, ys, Us[:,0]*arrow_coeff, Us[:,1]*arrow_coeff, color='black')
im = ax[1].scatter(xs,ys,c=Us[:,2]*1000,label='z displacement', vmax=vmax, vmin=-vmax, cmap='seismic') 
cbar = plt.colorbar(im)
cbar.set_label(r'd$_z$ (mm) ')
ax[1].set_title('displacement without noise')
ax[1].set_ylabel('y (m)')
ax[1].set_xlabel('x (x)')
ax[1].locator_params(nbins=6,axis='x')
ax[1].set_xlim(-2000,2000)
ax[1].set_ylim(-2000,2000)

ax[2].quiver(xs, ys, Us_noisy[:,0]*arrow_coeff, Us_noisy[:,1]*arrow_coeff, color='black')
im = ax[2].scatter(xs,ys,c=Us_noisy[:,2]*1000,label='z displacement', vmax=vmax, vmin = -vmax, cmap='seismic') 
cbar = plt.colorbar(im)
cbar.set_label(r'd$_z$ (mm) ')
ax[2].set_title('displacement with noise')
ax[2].set_ylabel('y (m)')
ax[2].set_xlabel('x (x)')
ax[2].locator_params(nbins=6,axis='x')
ax[2].set_xlim(-2000,2000)
ax[2].set_ylim(-2000,2000)

fig.subplots_adjust(wspace=0.5)
```

```python
data_number = len(Us[:,0])*3 #number of receivers *3 
source_number = ps.size
l = E * v / ((1 + v) * (1 - 2 * v))
m = E / (2 * (1 + v))    
disp_obs = Us_noisy.reshape((data_number,1))
G = disp2load.build_G(rs,xs,ys,l,m)
laplacian = disp2load.build_laplacian(rs)
TV = np.sqrt(np.abs(laplacian))


lambs = np.logspace(-19,0,80)
sigmas = np.logspace(-6,3,50) #looping over sigmas to also find the 
gaussians = np.zeros((source_number,source_number,len(sigmas)))
for i,sigma in enumerate(sigmas):
    gaussians[:,:,i] = disp2load.build_gaussian_inv(rs,sigma=sigma,normalized=True)
norm_laplacian_ps = np.zeros(len(lambs)) #calcul de la norme des solutions
norm_residuals_laplacian = np.zeros(len(lambs)) #calcul de la norme des résidus
norm_diff_ps_laplacian = np.zeros(len(lambs)) #calcul de la norme des résidus
norm_TV_ps = np.zeros(len(lambs))
norm_residuals_TV = np.zeros(len(lambs))
norm_diff_ps_TV = np.zeros(len(lambs)) 
norm_gaussian_ps = np.zeros((len(sigmas),len(lambs)))
norm_residuals_gaussian = np.zeros((len(sigmas),len(lambs)))
norm_diff_ps_gaussian = np.zeros((len(sigmas),len(lambs))) 


for i in trange(len(lambs)):
    lamb = lambs[i]
    ##### LAPLACIAN COMPUTAITONS #######
    ps_laplacian = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=lamb,Cm1=laplacian,G=G) #precomputed G and lalaplcian to save time
    ps_laplacian_reshaped = ps_laplacian.reshape((source_number,1))
    norm_laplacian_ps[i] = np.sqrt(np.sum((laplacian@ps_laplacian_reshaped)**2)) 
    disp_cal = G@ps_laplacian_reshaped
    norm_residuals_laplacian[i] = np.sqrt(np.sum((disp_obs-disp_cal)**2))
    norm_diff_ps_laplacian[i] =  np.sqrt(np.sum((ps-ps_laplacian)**2))    
    ###### TV COMPUTAITONS #######
    ps_TV = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='TV', lamb=lamb, epsilon=1e-0, gamma_coeff=0.1,G=G) #for TV need to use very low values of lambda
    ps_TV_reshaped = ps_TV.reshape((source_number,1))
    norm_TV_ps[i] = np.sqrt(np.sum((TV@ps_TV_reshaped)**2)) 
    disp_cal = G@ps_TV_reshaped
    norm_residuals_TV[i] = np.sqrt(np.sum((disp_obs-disp_cal)**2))
    norm_diff_ps_TV[i] =  np.sqrt(np.sum((ps-ps_TV)**2))   
    ##### GAUSSIAN COMPUTATIONS ######    
    for j,sigma in enumerate(sigmas):
        gaussian = gaussians[:,:,j]
        ps_gaussian = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=lamb,G=G,Cm1=gaussian)
        ps_gaussian_reshaped = ps_gaussian.reshape((source_number,1))
        norm_gaussian_ps[j,i] = np.sqrt(np.sum((gaussian@ps_gaussian_reshaped)**2)) 
        disp_cal = G@ps_gaussian_reshaped
        norm_residuals_gaussian[j,i] = np.sqrt(np.sum((disp_obs-disp_cal)**2))
        norm_diff_ps_gaussian[j,i] =  np.sqrt(np.sum((ps-ps_gaussian)**2))   
    

#### last computation ###
### laplacian finding the best value of lambda and the one predicted by the ;curve 
idx_lambda_optimum_laplacian = np.argmin(norm_diff_ps_laplacian)
def second_deriv_Lcurve(norm_regu,misfit,lambda_list,lambda_range):
    """Computes the second derivative of the L curve -> allows to detect the elbow of the L-curve"""
    sorted_indices = np.argsort(norm_regu)
    x_sorted = norm_regu[sorted_indices]
    y_sorted = misfit[sorted_indices]
    lambda_list_sorted = lambda_list[sorted_indices]
    idx2keep = np.where((lambda_list_sorted>=lambda_range[0])&(lambda_list_sorted<=lambda_range[1]))[0]
    x_sorted_cropped = x_sorted[idx2keep]
    y_sorted_cropped = y_sorted[idx2keep]
    lambda_list_cropped = lambda_list_sorted[idx2keep]
    dy_dx = np.gradient(y_sorted_cropped, x_sorted_cropped)
    d2y_dx2 = np.gradient(dy_dx, x_sorted_cropped)
    return x_sorted_cropped, d2y_dx2, lambda_list_cropped

x_laplacian, d2_laplacian, lambda_list_laplacian = second_deriv_Lcurve(norm_laplacian_ps,norm_residuals_laplacian,lambs,[1e-8,1e-5]) #comp
idx_lambda_Lcurve_laplacian = np.argmax(d2_laplacian) #we can see that we don't find quite the same value of lambda between the truth and the Lcurve, but its close enough with the noise level we have
print(fr'relative error on $\lambda$ with Laplacian approach : {100*np.abs(lambs[idx_lambda_optimum_laplacian]-lambda_list_laplacian[idx_lambda_Lcurve_laplacian])/np.min([lambs[idx_lambda_optimum_laplacian],lambda_list_laplacian[idx_lambda_Lcurve_laplacian]])} (%)')

### TV
idx_lambda_optimum_TV = np.argmin(norm_diff_ps_TV)
x_TV, d2_TV, lambda_list_TV = second_deriv_Lcurve(norm_TV_ps,norm_residuals_TV,lambs,[1e-15,1e-11]) #comp
idx_lambda_Lcurve_TV = np.argmax(d2_TV) #we can see that we don't find quite the same value of lambda between the truth and the Lcurve, but its close enough with the noise level we have
print(fr'relative error on $\lambda$ with TV approach : {100*np.abs(lambs[idx_lambda_optimum_laplacian]-lambda_list_TV[idx_lambda_Lcurve_TV])/np.min([lambs[idx_lambda_optimum_laplacian],lambda_list_TV[idx_lambda_Lcurve_TV]])} (%)')

### Gaussian
idx_sigma_correct = np.argmin(np.abs(sigmas-0.7543120063354622))
#what value of sigma should be used to have the best estimate from the L-curve
idx_lambda_optimum_gaussian = np.argmin(norm_diff_ps_gaussian[idx_sigma_correct,:]) #finding the index of lambda for the chosen standard deviation
x_gaussian, d2_gaussian, lambda_list_gaussian = second_deriv_Lcurve(norm_gaussian_ps[idx_sigma_correct,:],
                                                                    norm_residuals_gaussian[idx_sigma_correct,:],lambs,[1e-19,1e-14]) #comp
idx_lambda_Lcurve_gaussian = np.argmax(d2_gaussian) #we can see that we don't find quite the same value of lambda between the truth and the Lcurve, but its close enough with the noise level we have

##### plotting the result

fig, ax = plt.subplots(3, 3,figsize=(14,9))
######## LAPLACIAN #### 
ax[0,0].set_title('Laplacian approach')
ax[0,0].loglog(lambs, norm_diff_ps_laplacian,linewidth=0, marker='.')
ax[0,0].loglog(lambs[idx_lambda_optimum_laplacian], norm_diff_ps_laplacian[idx_lambda_optimum_laplacian],linewidth=0, marker='.',color='red')
ax[0,0].set_xlabel(r'$\lambda$')
ax[0,0].set_ylabel(r'$|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$') #ps is the true load distribution 
im = ax[1,0].scatter(norm_laplacian_ps, norm_residuals_laplacian,c=lambs, cmap='turbo', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[1,0].set_xlabel(r'$|| \nabla \mathbf{p} ||^2_2$')
ax[1,0].set_ylabel(r'$|| \mathbf{G p} - \mathbf{d} ||_2^2$')
im = ax[2,0].scatter(x_laplacian, d2_laplacian,c=lambda_list_laplacian, cmap='viridis', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[2,0].set_xlabel(r'$|| \nabla \mathbf{p} ||^2_2$')
ax[2,0].set_ylabel(r'$ \frac{ \partial^2 || \mathbf{G p} - \mathbf{d} ||_2^2 } { {|| \nabla \mathbf{p} ||^2_2}^2} $')
######## TV ###################### -> no real possiblity to go for the L curve 
ax[0,1].set_title('TV approach')
ax[0,1].loglog(lambs, norm_diff_ps_TV,linewidth=0, marker='.')
ax[0,1].loglog(lambs[idx_lambda_optimum_TV], norm_diff_ps_TV[idx_lambda_optimum_TV],linewidth=0, marker='.',color='red')
ax[0,1].set_xlabel(r'$\lambda$')
ax[0,1].set_ylabel(r'$|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$') #ps is the true load distribution 
im = ax[1,1].scatter(norm_TV_ps, norm_residuals_TV,c=lambs, cmap='turbo', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[1,1].set_xlabel(r'$|| \nabla \mathbf{p} ||_1$')
ax[1,1].set_ylabel(r'$|| \mathbf{G p} - \mathbf{d} ||_2^2$')
im = ax[2,1].scatter(x_TV, d2_TV,c=lambda_list_TV, cmap='viridis', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[2,1].set_xlabel(r'$|| \nabla \mathbf{p} ||_1$')
ax[2,1].set_ylabel(r'$  \frac{ \partial^2 || \mathbf{G p} - \mathbf{d} ||_2^2 } {|| \nabla \mathbf{p} ||_1} $')

######### GAUSSIAN ###########
ax[0,2].set_title('Gaussian approach')
im = ax[0,2].pcolormesh(lambs,sigmas,norm_diff_ps_gaussian)
cbar = plt.colorbar(im)
cbar.set_label(r'$|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$')
ax[0,2].set_xlabel(r'$\lambda$')
ax[0,2].set_ylabel(r'$\sigma$') #ps is the true load distribution 
ax[0,2].set_xscale('log')
ax[0,2].set_yscale('log')
im = ax[1,2].scatter(norm_gaussian_ps[idx_sigma_correct,:], norm_residuals_gaussian[idx_sigma_correct,:],c=lambs, cmap='turbo', norm=matplotlib.colors.LogNorm(),label=fr'$\sigma$={sigmas[idx_sigma_correct]}')
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[1,2].set_xlabel(r'$|| \mathcal{G} \mathbf{p} ||^2_2$')
ax[1,2].set_ylabel(r'$|| \mathbf{G p} - \mathbf{d} ||_2^2$')
ax[1,2].legend()
im = ax[2,2].scatter(x_gaussian, d2_gaussian,c=lambda_list_gaussian, cmap='viridis', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[2,2].set_xlabel(r'$|| \mathcal{G} \mathbf{p} ||^2_2$')
ax[2,2].set_ylabel(r'$ \frac { \partial^2 || \mathbf{G p} - \mathbf{d} ||_2^2 }{{|| \mathcal{G} \mathbf{p} ||^2_2}^2}$')

fig.subplots_adjust(wspace=0.5, hspace=0.3)
```

With the increased noise level, it appears that deriving $\lambda$ from the L-curve using its second derivative is more difficult.


```python
ps_large_sigma = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=1e-16,
                                  G=G,Cm1='gaussian',sigma=1e1)

ps_large_lambda = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=1e-12,
                                  G=G,Cm1='gaussian',sigma=1e-1)

ps_optimum_pair = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=lambs[idx_lambda_optimum_gaussian],
                                  G=G,Cm1='gaussian',sigma=sigmas[idx_sigma_correct])

idx_selec = np.argmin(np.abs(1e-1-sigmas))
ps_good_pair = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='linear', lamb=lambs[np.argmin(norm_diff_ps_gaussian[idx_selec,:])],
                                  G=G,Cm1='gaussian',sigma=sigmas[idx_selec])

fig, ax = plt.subplots(1,4,figsize=(14,2))
#suing the same scale for the different schemes 
vmin = np.min([ps_large_lambda,ps_large_sigma,ps_optimum_pair,ps_good_pair])
vmax = np.max([ps_large_lambda,ps_large_sigma,ps_optimum_pair,ps_good_pair])

im = ax[0].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_optimum_pair, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0].set_xlabel('x (m)')
ax[0].set_ylabel('y (m)')
ax[0].set_title(fr'Optimum pair : $\lambda$={"{:.1e}".format(lambs[idx_lambda_optimum_gaussian])}, $\sigma$={"{:.1e}".format(sigmas[idx_sigma_correct])}')

im = ax[1].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_good_pair, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[1].set_xlabel('x (m)')
ax[1].set_ylabel('y (m)')
ax[1].set_title(fr'Good pair : $\lambda$={"{:.1e}".format(lambs[np.argmin(norm_diff_ps_gaussian[idx_selec,:])])}, $\sigma$={"{:.1e}".format(sigmas[idx_selec])}')

im = ax[2].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_large_sigma, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[2].set_xlabel('x (m)')
ax[2].set_ylabel('y (m)')
ax[2].set_title(fr'Too large $\sigma$ : $\lambda$=1e-17, $\sigma$=1e1')

im = ax[3].pcolormesh(np.linspace(-1750,1750,8),np.linspace(-1750,1750,8),ps_large_lambda, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[3].set_xlabel('x (m)')
ax[3].set_ylabel('y (m)')
ax[3].set_title(fr'Too large $\lambda$ : $\lambda$=1e-12, $\sigma$=1e-1')

fig.subplots_adjust(wspace=1, hspace=0.3)
```

While the load distribution inverted using the optimum pair of parameters is fairly close to the actual load distribution, the load distribution inverted using a slightly lower value of $\sigma$ and its associated optimum $\lambda$ is completely wrong. This shows that in order for the inversion to be successfull with large noise, one must use the best $\sigma$. We can see that when $\sigma$ is too large, the inversion is unstable and leads to a characteristic null load distribution, and we previously noticed that the stability of the estimated value of $\sigma$ from L-curve can also be used to know wether the estimated value of $\sigma$ too small.   

Since we can definetely notice the unstable behavior for large $\sigma$, we only need to verify that the values of $\lambda$ derived from the L-curve remain constant for small values of $\sigma$, and that a perculiar behavior occurs once the value of $\sigma$ is optimum.

```python
lambdas_true = np.zeros(len(sigmas))
lambdas_Lcurve = np.zeros(len(sigmas))
misfit4optimum = np.zeros(len(sigmas))

for i, sigma in enumerate(sigmas):
    lambdas_true[i] = lambs[np.argmin(norm_diff_ps_gaussian[i,:])]
    misfit4optimum[i] = np.min(norm_diff_ps_gaussian[i,:]) #also keeping that to be able to tell if the relative error of lambda is relevant 
    x, d2, lambda_list = second_deriv_Lcurve(norm_gaussian_ps[i,:],
                                            norm_residuals_gaussian[i,:],lambs,[1e-18,1e-15])
    lambdas_Lcurve[i] = lambda_list[np.argmax(d2)]  
    
relative_error = 100*np.abs(lambdas_Lcurve-lambdas_true)/np.min([lambdas_Lcurve,lambdas_true],axis=0)

print(r'The lowest error between the true and inverted load distributions is obtained for $\sigma$ = ',sigmas[np.argmin(misfit4optimum)])
print(fr'For sigma = {sigmas[np.argmin(misfit4optimum)]}, the relative error of lambda is {relative_error[np.argmin(misfit4optimum)]} %')

###

fig, ax = plt.subplots(1,3, figsize=(15,4))
ax[0].scatter(sigmas,lambdas_true, label='Optimum')
ax[0].scatter(sigmas,lambdas_Lcurve, label='Lcurve')
ax[0].set_title(r'$\lambda$ vs $\sigma$')
ax[0].set_xlabel(r'$\sigma$')
ax[0].set_xscale('log')
ax[0].set_ylabel(r'$\lambda$')
ax[0].set_yscale('log')
ax[0].legend()

ax[1].scatter(sigmas,relative_error) #plotting the relative error between the value of lambda found with l-curve and the optimum one 
ax[1].set_ylabel(r'relative error $|\lambda_{L-curve}-\lambda_{true}|/ \lambda_{true}$ (%)')
ax[1].set_xlabel(r'$\sigma$')
ax[1].set_xscale('log')
ax[1].set_title(r'Relative error for $\lambda$')
ax[1].set_yscale('log')

ax[2].scatter(sigmas,misfit4optimum)
ax[2].set_xlabel(r'$\sigma$')
ax[2].set_ylabel(r'$|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$')
ax[2].set_xscale('log')
ax[2].set_title('Misfit for optimum lambda')
ax[2].set_yscale('log')
fig.subplots_adjust(wspace=0.3)
```

We can see that with this higher noise level, while the predicted value of lambda with the L-curve remains constant for low $\sigma$, it is significantly different than the optimum one (about 300 % relative error), even for small values of $\sigma$. In addition, in this example the best $\sigma$ is about 0.75, and for this standard deviation, the relative error between the optimum $\lambda$ and the $\lambda$ derived from the L-curve is quite large (800 %).

Based on these observations it appears that for large noise level, one cannot guess $\sigma$ based on the last value of $\lambda$ that is identical to the previous ones. However, one can see that as before, the L-curve starts predicting larger values of $\lambda$ than for small $\sigma$, while we know that the optimum value of $\lambda$ remains constant for low values of $\sigma$ and decreases for large values of $\sigma$ (until unstability). Therefore, we may still be able to guess the optimum $\sigma$ as being the value above which the inversion is unstable and below which the $\lambda$ derived from L-curve remains constant.

While it appears possible to accurately constaint $\sigma$ with this approach, the value of $\lambda$ estimated through the L-curve may actually not be suited and the result of the inversion may be wrong. Because all the estimations rely on the way we estimate the corner of the L-curve (second derivative of $|| \mathbf{G p} - \mathbf{d} ||_2^2$ with respect to $||\mathcal{G} \mathbf{p} ||_2^2$); we may find completely different results if we determined the corner with a different approach. 

Below we display the result of the inversion for the Laplacian, the TV and the Gaussian, using either the optimum value of $\lambda$ or the value of $\lambda$ derived from the L-curve.

We can see that with large noise, the total variation starts to perform poorly, as the spacial extent of the sources is not well constrained. On the other hand, one can see that the Gaussian and Laplacian perform similarly, with the Gaussian performing slightly better than the Laplacian (no negative load in the central region).

When using the values of $\lambda$ derived from the L-curve, Total Variation and Gaussian approaches perform poorly, as the estimated value of $\lambda$ is very different from the optimum, and that the result of the inversion is very dependant on the value of $\lambda$ used. Meanwhile, even though the estimated value of $\lambda$ is also wrongly estimated for the Laplacian approach; it still performs quite well.

From this observation it seems that Laplacian is to favor for large noise level, however Gaussian approach may still be of use for large number of stations if we use another method to find the corner of the L-curve.

The advantage of these non-linear approach is that they allow to add constraints on the values of the inverted parameters. In order to determine the best non-linear approch (fastest and closest to the true load distribution), we will perform a time convergence analysis.

### Improving the result with constraints

We now try to improve the result of the inversion by adding constraints on the load of certain regions of the map.

```python
constraint = np.empty((8,8)) 
constraint[:,:] = np.nan 
constraint[0:4,2:4] = 0
constraint[2:4,0:4] = 0

norm_diff_ps_laplacian_constrained = np.zeros(len(lambs)) 
norm_diff_ps_gaussian_constrained = np.zeros((len(sigmas),len(lambs))) 

for i in trange(len(lambs)):
    lamb = lambs[i]
    ##### LAPLACIAN COMPUTAITONS #######
    ps_laplacian_constrained = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='lbfgs_fast', lamb=lamb,Cm1=laplacian,G=G,constraint=constraint,
                                      epsilon=1e-35,maxit=1e4) #precomputed G and lalaplcian to save time
    norm_diff_ps_laplacian_constrained[i] =  np.sqrt(np.sum((ps-ps_laplacian_constrained)**2))    
    ##### GAUSSIAN COMPUTATIONS ######    
    for j,sigma in enumerate(sigmas):
        gaussian = gaussians[:,:,j]
        ps_gaussian_constrained = disp2load.disp2load(E,v,rs,xs,ys,Us_noisy,mode='lbfgs_fast', lamb=lamb,G=G,Cm1=gaussian,constraint=constraint,
                                         epsilon=1e-35,maxit=1e4)
        norm_diff_ps_gaussian_constrained[j,i] =  np.sqrt(np.sum((ps-ps_gaussian_constrained)**2))   
    
#### last computation ###
idx_lambda_optimum_laplacian_constrained = np.argmin(norm_diff_ps_laplacian)
### 
idx_sigma_optimum_gaussian_constrained,idx_lambda_optimum_gaussian_constrained = np.where(norm_diff_ps_gaussian_constrained==np.min(norm_diff_ps_gaussian_constrained))

print(r"The optimal value of $\lambda$ with Laplacian and constraint is :", lambs[idx_lambda_optimum_laplacian_constrained])
print(r"The optimal value of $\sigma$ and $\lambda$ with Gaussian and constraint are :", sigmas[idx_sigma_optimum_gaussian_constrained] ,' and ',lambs[idx_lambda_optimum_gaussian_constrained], 'respectively.')
```

We find that the new optimal values of $\sigma$ and $\lambda$ are identical to the ones without constraint.

### Simple load distrbution convergence time analysis for different inversion approaches

```python
#plotting the error and inverison time for various inversion scheme  précalcule à chaque fois G et le laplacien
#pour voir le temps de calcul...  -> utilise laplacien = np zeros  pour aller plus vite et ne pas faire de 
E = 75e9
v = 0.25
l = E * v / ((1 + v) * (1 - 2 * v))
m = E / (2 * (1 + v))

ps = np.zeros((2,2))
ps[0:1,0:1] = 10*9.81*1000 #we just put this quarter of the load distribution = 0 
rs = utils.create_source_mesh(0,1000,0,1000,ps) 
seed_value = 42
np.random.seed(seed_value)
xs = np.random.rand(70)*1000 #we keeping the same receivers for the different load distribution sizes
ys = np.random.rand(70)*1000
G = disp2load.build_G(rs,xs,ys,l,m)
ps = ps.reshape((ps.size,1))
Us = G@ps #should be of size 150, -> 3 disp per station
Us = Us.reshape((len(xs),3)) #we reshaped to the correct shape to be accepted by the inversion function

#number of pixel width of the load distribution, should be even  and we don't use any regularization so we use Cm1 = np.zeros()
width = np.arange(2,14,2)
time_linear = np.zeros(len(width))
time_steepest = np.zeros(len(width))
time_nlcg = np.zeros(len(width))
time_nlcg_fast = np.zeros(len(width))
time_lbfgs = np.zeros(len(width))
time_lbfgs_fast = np.zeros(len(width))
error_linear = np.zeros(len(width))
error_lbfgs = np.zeros(len(width))
error_lbfgs_fast = np.zeros(len(width))
error_nlcg = np.zeros(len(width))
error_nlcg_fast = np.zeros(len(width))
error_steepest = np.zeros(len(width))
for i in range(len(width)):
    ps = np.zeros((width[i],width[i]))
    ps[0:int(width[i])//2,0:int(width[i])//2] = 10*9.81*1000 #we just put this quarter of the load distribution = 0 
    rs = utils.create_source_mesh(0,1000,0,1000,ps)
    G = disp2load.build_G(rs,xs,ys,l,m) #now we have G we use it to invert the laod
    Cm1 = np.zeros((ps.size,ps.size))
    #####
    t1 = time.time()
    ps_linear = disp2load.disp2load(E,v,rs,xs,ys,Us,mode='linear', lamb=0, epsilon=1e-35, gamma_coeff=0.1,Cm1=Cm1,
                                      sigma=None,constraint=None, maxit=1e4, verbose=0, logpath=None)
    error_linear[i] = np.linalg.norm(ps-ps_linear)
    time_linear[i] = time.time()-t1
    ####
    t1 = time.time()
    ps_nlcg = disp2load.disp2load(E,v,rs,xs,ys,Us,mode='nlcg', lamb=0, epsilon=1e-35, gamma_coeff=0.1,Cm1=Cm1,
                                      sigma=None,constraint=None, maxit=1e4, verbose=0, logpath=None)
    error_nlcg[i] = np.linalg.norm(ps-ps_nlcg)
    time_nlcg[i] = time.time()-t1
    ####
    t1 = time.time()
    ps_nlcg_fast = disp2load.disp2load(E,v,rs,xs,ys,Us,mode='nlcg_fast', lamb=0, epsilon=1e-35, gamma_coeff=0.1,Cm1=Cm1,
                                      sigma=None,constraint=None, maxit=1e4, verbose=0, logpath=None)
    error_nlcg_fast[i] = np.linalg.norm(ps-ps_nlcg_fast)
    time_nlcg_fast[i] = time.time()-t1
    ####
    t1 = time.time()
    ps_steepest = disp2load.disp2load(E,v,rs,xs,ys,Us,mode='steepest', lamb=0, epsilon=1e-35, gamma_coeff=0.1,Cm1=Cm1,
                                      sigma=None,constraint=None, maxit=1e4, verbose=0, logpath=None)
    error_steepest[i] = np.linalg.norm(ps-ps_steepest)
    time_steepest[i] = time.time()-t1
    ####
    t1 = time.time()
    ps_lbfgs = disp2load.disp2load(E,v,rs,xs,ys,Us,mode='lbfgs', lamb=0, epsilon=1e-35, gamma_coeff=0.1,Cm1=Cm1,
                                      sigma=None,constraint=None, maxit=1e4, verbose=0, logpath=None)
    error_lbfgs[i] = np.linalg.norm(ps-ps_lbfgs)
    time_lbfgs[i] = time.time()-t1
    ####
    t1 = time.time()
    ps_lbfgs_fast = disp2load.disp2load(E,v,rs,xs,ys,Us,mode='lbfgs_fast', lamb=0, epsilon=1e-35, gamma_coeff=0.1,Cm1=Cm1,
                                      sigma=None,constraint=None, maxit=1e4, verbose=0, logpath=None)
    error_lbfgs_fast[i] = np.linalg.norm(ps-ps_lbfgs_fast)
    time_lbfgs_fast[i] = time.time()-t1
    
###plotting the result of the inversion vs the truth
d = 1000/width[i]
vmax = np.max(Us[:,2])

fig, ax = plt.subplots(1,2, figsize=(14,5))
# im = ax[0].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps)
# cbar = plt.colorbar(im)
# ax[0].scatter(xs,ys,c=Us[:,2],vmax=vmax, vmin=-vmax, cmap='seismic')
# ax[0].set_title('Original distribution')

ax[1].loglog(width**2,time_linear,label='Linear') #width**2 = source number  
ax[1].loglog(width**2,time_lbfgs,label='lbfgs')
ax[1].loglog(width**2,time_lbfgs_fast,label='lbfgs_fast')
ax[1].loglog(width**2,time_nlcg,label='nlcg')
ax[1].loglog(width**2,time_nlcg_fast,label='nlcg_fast')
ax[1].loglog(width**2,time_steepest,label='steepest')
ax[1].legend()
ax[1].set_ylabel('time (s)')
ax[1].set_xlabel('source number')
ax[1].set_title('Convergence time of different approaches')

ax[0].loglog(width**2,error_linear,label='Linear') #width**2 = source number  
ax[0].loglog(width**2,error_lbfgs,label='lbfgs')
ax[0].loglog(width**2,error_lbfgs_fast,label='lbfgs_fast')
ax[0].loglog(width**2,error_nlcg,label='nlcg')
ax[0].loglog(width**2,error_nlcg_fast,label='nlcg_fast')
ax[0].loglog(width**2,error_steepest,label='steepest')
ax[0].legend()
ax[0].set_ylabel('L2 error')
ax[0].set_xlabel('source number')
ax[0].set_title('L2 norm on the residual for different approaches')


fig, ax = plt.subplots(2,3,figsize=(14,5))
vmin = np.min([ps_linear,ps_lbfgs,ps_lbfgs_fast,ps_nlcg,ps_steepest,ps_nlcg_fast])
vmax = np.max([ps_linear,ps_lbfgs,ps_lbfgs_fast,ps_nlcg,ps_steepest,ps_nlcg_fast])

im = ax[0,0].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps_linear,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[0,0].set_title('Linear')

im = ax[1,0].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps_steepest,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[1,0].set_title('Steepest')

im = ax[0,1].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps_lbfgs,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[0,1].set_title('lbfgs')

im = ax[1,1].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps_lbfgs_fast,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[1,1].set_title('lbfgs_fast')

im = ax[0,2].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps_nlcg,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[0,2].set_title('nlcg')

im = ax[1,2].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps_nlcg_fast,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[1,2].set_title('nlcg_fast')

fig.subplots_adjust(wspace=0.3, hspace=0.3)
```

We can see that as observed previously, the NLCG fast approach performs the worse in terms of error, however it is as fast as the linear inversion. Overall among the different inversion approaches, the L-BFGS seems to perform the best as it is the the closest to the actual load distribution and is remarkably fast. While our implementation of the L-BFGS algorithm has about the same error as the one from scipy, it is noticeably slower than the one from scipy. For that reason we may use the scipy implementation for large computations.

```python
#we adding constraints, knowing that we don't add any noise

#number of pixel width of the load distribution, should be even  and we don't use any regularization so we use Cm1 = np.zeros()
width = np.arange(2,14,2)
time_linear = np.zeros(len(width))
time_steepest = np.zeros(len(width))
time_nlcg = np.zeros(len(width))
time_nlcg_fast = np.zeros(len(width))
time_lbfgs = np.zeros(len(width))
time_lbfgs_fast = np.zeros(len(width))
error_linear = np.zeros(len(width))
error_lbfgs = np.zeros(len(width))
error_lbfgs_fast = np.zeros(len(width))
error_nlcg = np.zeros(len(width))
error_nlcg_fast = np.zeros(len(width))
error_steepest = np.zeros(len(width))
for i in range(len(width)):
    ### creating the constraint to be used for the 
    constraint = np.empty((width[i],width[i]))
    constraint[:] = np.nan
    constraint[int(width[i])//2:,:] = 0 #putting the 0 constraint for the 
    #####
    ps = np.zeros((width[i],width[i]))
    ps[0:int(width[i])//2,0:int(width[i])//2] = 10*9.81*1000 #we just put this quarter of the load distribution = 0 
    rs = utils.create_source_mesh(0,1000,0,1000,ps)
    G = disp2load.build_G(rs,xs,ys,l,m) #now we have G we use it to invert the laod
    Cm1 = np.zeros((ps.size,ps.size))
    #####
    t1 = time.time()
    ps_nlcg = disp2load.disp2load(E,v,rs,xs,ys,Us,mode='nlcg', lamb=0, epsilon=1e-35, gamma_coeff=0.1,Cm1=Cm1,
                                      sigma=None,constraint=constraint, maxit=1e4, verbose=0, logpath=None)
    error_nlcg[i] = np.linalg.norm(ps-ps_nlcg)
    time_nlcg[i] = time.time()-t1
    ####
    t1 = time.time()
    ps_steepest = disp2load.disp2load(E,v,rs,xs,ys,Us,mode='steepest', lamb=0, epsilon=1e-35, gamma_coeff=0.1,Cm1=Cm1,
                                      sigma=None,constraint=constraint, maxit=1e4, verbose=0, logpath=None)
    error_steepest[i] = np.linalg.norm(ps-ps_steepest)
    time_steepest[i] = time.time()-t1
    ####
    t1 = time.time()
    ps_lbfgs = disp2load.disp2load(E,v,rs,xs,ys,Us,mode='lbfgs', lamb=0, epsilon=1e-35, gamma_coeff=0.1,Cm1=Cm1,
                                      sigma=None,constraint=constraint, maxit=1e4, verbose=0, logpath=None)
    error_lbfgs[i] = np.linalg.norm(ps-ps_lbfgs)
    time_lbfgs[i] = time.time()-t1
    ####
    t1 = time.time()
    ps_lbfgs_fast = disp2load.disp2load(E,v,rs,xs,ys,Us,mode='lbfgs_fast', lamb=0, epsilon=1e-35, gamma_coeff=0.1,Cm1=Cm1,
                                      sigma=None,constraint=constraint, maxit=1e4, verbose=0, logpath=None)
    error_lbfgs_fast[i] = np.linalg.norm(ps-ps_lbfgs_fast)
    time_lbfgs_fast[i] = time.time()-t1
    
###plotting the result of the inversion vs the truth
fig, ax = plt.subplots(1,2, figsize=(14,5))
ax[1].loglog(width**2,time_lbfgs,label='lbfgs')
ax[1].loglog(width**2,time_lbfgs_fast,label='lbfgs_fast')
ax[1].loglog(width**2,time_nlcg,label='nlcg')
ax[1].loglog(width**2,time_steepest,label='steepest')
ax[1].legend()
ax[1].set_ylabel('time (s)')
ax[1].set_xlabel('source number')
ax[1].set_title('Convergence time of different approaches')

ax[0].loglog(width**2,error_lbfgs,label='lbfgs')
ax[0].loglog(width**2,error_lbfgs_fast,label='lbfgs_fast')
ax[0].loglog(width**2,error_nlcg,label='nlcg')
ax[0].loglog(width**2,error_steepest,label='steepest')
ax[0].legend()
ax[0].set_ylabel('L2 error')
ax[0].set_xlabel('source number')
ax[0].set_title('L2 norm on the residual for different approaches')


fig, ax = plt.subplots(2,3,figsize=(14,5))
d = 1000/width[i] #step in x and y

vmin = np.min([ps,ps_lbfgs,ps_lbfgs_fast,ps_nlcg,ps_steepest])
vmax = np.max([ps,ps_lbfgs,ps_lbfgs_fast,ps_nlcg,ps_steepest])

im = ax[0,0].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[0,0].set_title('Original')

im = ax[1,2].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps_steepest,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[1,2].set_title('Steepest')

im = ax[0,1].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps_lbfgs,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[0,1].set_title('lbfgs')

im = ax[1,1].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps_lbfgs_fast,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[1,1].set_title('lbfgs_fast')

im = ax[0,2].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),ps_nlcg,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[0,2].set_title('nlcg')

im = ax[1,0].pcolormesh(np.linspace(0+d/2,1000-d/2,int(width[i])),np.linspace(0+d/2,1000-d/2,int(width[i])),constraint,vmax=vmax,vmin=vmin)
cbar = plt.colorbar(im)
ax[1,0].set_title('constraint')

fig.subplots_adjust(wspace=0.3, hspace=0.3)
```

We can see that the NLCG doesn't work well with the constraint, while the Steepest descent and the LBFGS approaches work very well with the constraint. In addition, one can see that our implementation of the LBFGS algorithm is more stable than the one from scipy.

### Inversion a of a complex load distribution with small noise and with/without constraints

Now we complexify the shape of the source to see the performance of the inversion. Because we've seen that overall the linear inversion as well as the LBFGS perform well (small error in the inversion and short computaiton time), we perform the inversion of the complex load distribution with these schemes. In addition, we saw that for low noise level, the L-curve was able to fairly well assess the correct value of $\lambda$ with the laplacian regularization, however it struggles with large noise level, therefore we will look at the result of the inversion using both the value of $\lambda$ derived from the L-curve and its optimum.

```python
#we recreate the laod distribution of the lake and we compute the displacements, add noise and then invert for the load distribution with linear and gaussian approaches
p = 1150*9.81*1 #rho g z
s = pd.read_csv(f'true_source.csv') #file containing the coordinates of the contour of the lake
lat_coordinates = s['lat'].values
lon_coordinates = s['lon'].values
x_coordinates, y_coordinates = utils.latlon2xy(lat_coordinates, lon_coordinates, zone=11) #converting to cartesian coordinates
r = np.zeros((len(x_coordinates),2))
r[:,0] = x_coordinates
r[:,1] = y_coordinates
#We now have all the coordinates 
##################### approach with multiple small sources ##################################################
width = 70
ps = np.zeros((width,width)) #gonna create a mesh of 100*100 sources
#need to match the extent from the figure of D'Urso and Marmo to easily compare our results
xstart, ystart = utils.latlon2xy(40.5,-113.5,zone=11)
xend, yend = utils.latlon2xy(42,-111.5,zone=11)
dx = (xend-xstart)/width
dy = (yend-ystart)/width
dlon = 2/width
dlat = 1.5/width
rs = utils.create_source_mesh(xstart,xend,ystart,yend,ps) #sources should have part out of the domain in which we compute 
xr = np.linspace(xstart+dx/2,xend-dx/2,width)
yr = np.linspace(ystart+dy/2,yend-dy/2,width)
latr = np.linspace(40.5+dlat/2,42-dlat/2,width)
lonr = np.linspace(-113.5+dlon/2,-111.5-dlon/2,width)

for j in range(len(ps[:,0])):
    for i in range(len(ps[0,:])):
        result = utils.is_point_inside_polygon(r, xr[i], yr[j]) #we loop over the sources and tell if they are inside the polygon r
        if result == True:
            ps[j,i] = p  #if inside the lake then we use the load from 1m of water of rho = 1150 kg/m^3
```

```python
#now we have the load distribution for the 2 approaches we can compute the displacements and compare it to what was obtained by D'Urso and Marmo.
E = 75e9
v = 0.25
l = E * v / ((1 + v) * (1 - 2 * v))
m = E / (2 * (1 + v))

data_number = 1700 #gotta use a large number of stations to remain in an overdetermined problem
# xs = np.random.randint(xstart, high=xend, size=data_number) #defining the coordinates of the receivers
# ys = np.random.randint(ystart, high=yend, size=data_number)
# np.save('xs_salt.npy',xs)
# np.save('ys_salt.npy',ys)
# G = disp2load.build_G(rs,xs,ys,l,m)
# np.save(f'G_salt4inversion.npy', G)
xs, ys = np.load('xs_salt.npy'), np.load('ys_salt.npy')
lats, lons = utils.xy2latlon(xs,ys,zone=11)
G = np.load('G_salt4inversion.npy')

source_number = width**2
ps_reshaped = ps.reshape(source_number,1)
Us_multiple = G@ps_reshaped #returns a vector of size ny*nx*3
Us_multiple = Us_multiple.reshape((data_number,3))

#### adding noise
seed_value = 42
np.random.seed(seed_value)
Us_multiple_noisy = Us_multiple.copy()
Us_multiple_noisy[:,0:2] += np.random.normal( loc=0.0, scale=0.05*np.max(np.abs(Us_multiple[:,0:2])), size=Us_multiple[:,0:2].shape ) #pourrait faire noise plus ou moins fort sur les composantes -> par ex 
Us_multiple_noisy[:,2] += np.random.normal( loc=0.0, scale=0.05*np.max(np.abs(Us_multiple[:,2])), size=Us_multiple[:,2].shape ) #pourrait faire noise plus ou moins fort sur les composantes -> par ex 

### now plotting the resulting displacements 
vmax = np.max([np.abs(Us_multiple_noisy[:,2]),np.abs(Us_multiple[:,2])])*1000
arrow_coeff = 10000000

fig, ax = plt.subplots(1, 3,figsize=(15,3))
im = ax[0].pcolormesh(lonr,latr,ps)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0].scatter(lons,lats,marker='^',linewidths=1, edgecolors='black', color='white')
ax[0].set_xlabel('longitude (°)')
ax[0].set_ylabel('latitude (°)')
ax[0].set_xlim(-113.5,-111.5)
ax[0].set_ylim(40.5,42)


im = ax[1].scatter(lons,lats,c=Us_multiple[:,2]*1000,label='z displacement', vmax=vmax, vmin=-vmax, cmap='seismic',s=5) 
ax[1].quiver(lons, lats, Us_multiple[:,0]*arrow_coeff, Us_multiple[:,1]*arrow_coeff, color='black')
cbar = plt.colorbar(im)
cbar.set_label(r'd$_z$ (mm) ')
ax[1].set_title('displacement without noise')
ax[1].set_ylabel('latitude (°)')
ax[1].set_xlabel('longitude (°)')
ax[1].set_xlim(-113.5,-111.5)
ax[1].set_ylim(40.5,42)

im = ax[2].scatter(lons,lats,c=Us_multiple_noisy[:,2]*1000, vmax=vmax, vmin =-vmax, cmap='seismic',s=5) 
ax[2].quiver(lons, lats, Us_multiple_noisy[:,0]*arrow_coeff, Us_multiple_noisy[:,1]*arrow_coeff, color='black')
cbar = plt.colorbar(im)
cbar.set_label(r'd$_z$ (mm) ')
ax[2].set_title('displacement with noise')
ax[2].set_ylabel('latitude (°)')
ax[2].set_xlabel('longitude (°)')
ax[2].set_xlim(-113.5,-111.5)
ax[2].set_ylim(40.5,42)

fig.subplots_adjust(wspace=0.5)
```

```python
#now we perform the inversion with linear and LBFGS approahces and with and without constraint
laplacian = np.load('laplacian_salt.npy')

lambs = np.logspace(-8,3,70)

norm_laplacian_ps = np.zeros(len(lambs)) #calcul de la norme des solutions
norm_residuals_laplacian = np.zeros(len(lambs)) #calcul de la norme des résidus
norm_diff_ps_laplacian = np.zeros(len(lambs)) #calcul de la norme des résidus
misfit_horizontal = np.zeros(len(lambs))
misfit_vertical = np.zeros(len(lambs))

disp_obs = Us_multiple_noisy.reshape((data_number*3,1)) #in this case we set data number as the number of stations instead of the actual number of data

for i in trange(len(lambs)):
    lamb = lambs[i]
    ##### LAPLACIAN COMPUTAITONS #######
    ps_laplacian = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='linear', lamb=lamb,Cm1=laplacian,G=G) #precomputed G and lalaplcian to save time
    ps_laplacian_reshaped = ps_laplacian.reshape((source_number,1))
    norm_laplacian_ps[i] = np.sqrt(np.sum((laplacian@ps_laplacian_reshaped)**2)) 
    disp_cal = G@ps_laplacian_reshaped
    norm_residuals_laplacian[i] = np.sqrt(np.sum((disp_obs-disp_cal)**2))
    norm_diff_ps_laplacian[i] =  np.sqrt(np.sum((ps-ps_laplacian)**2))    
    
    Us_cal = disp_cal.reshape((1700,3)) #reshaping in the same shape as the Us_multiple_noisy  to more easily separate the displacements on the diff components 
    diff_e = np.mean(np.abs(Us_cal[:,0]-Us_multiple_noisy[:,0]))
    diff_n = np.mean(np.abs(Us_cal[:,1]-Us_multiple_noisy[:,0]))
    misfit_horizontal[i] = np.mean([diff_e, diff_n]) 
    misfit_vertical[i] = np.mean(np.abs(Us_cal[:,2]-Us_multiple_noisy[:,2]))

#getting the value of lambda from the L-curve 
idx_lambda_optimum_laplacian = np.argmin(norm_diff_ps_laplacian)
x_laplacian, d2_laplacian, lambda_list_laplacian = utils.second_deriv_Lcurve(norm_laplacian_ps,norm_residuals_laplacian,lambs,[1e-7,1e-3])
idx_lambda_Lcurve_laplacian = np.argmax(d2_laplacian)
print(fr'relative error on $\lambda$ with Laplacian approach : {100*np.abs(lambs[idx_lambda_optimum_laplacian]-lambda_list_laplacian[idx_lambda_Lcurve_laplacian])/np.min([lambs[idx_lambda_optimum_laplacian],lambda_list_laplacian[idx_lambda_Lcurve_laplacian]])} (%)')    
#we actually estimate very poorly the value of lambda with the L-curve in this case ...

### plotting the curves to see if the poor estimate of lambda is due to the automatic extraction or to the L-curve itself
fig, ax = plt.subplots(1, 3,figsize=(15,3))
######## LAPLACIAN #### 
# ax[0,0].set_title('Laplacian approach')
ax[0].loglog(lambs, norm_diff_ps_laplacian,linewidth=0, marker='.')
ax[0].loglog(lambs[idx_lambda_optimum_laplacian], norm_diff_ps_laplacian[idx_lambda_optimum_laplacian],linewidth=0, marker='.',color='red')
ax[0].set_xlabel(r'$\lambda$')
ax[0].set_ylabel(r'$|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$') #ps is the true load distribution 
im = ax[1].scatter(norm_laplacian_ps, norm_residuals_laplacian,c=lambs, cmap='turbo', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[1].set_xlabel(r'$|| \nabla \mathbf{p} ||^2_2$')
ax[1].set_ylabel(r'$|| \mathbf{G p} - \mathbf{d} ||_2^2$')
im = ax[2].scatter(x_laplacian, d2_laplacian,c=lambda_list_laplacian, cmap='viridis', norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(im)
cbar.set_label(r'$\lambda$')
ax[2].set_xlabel(r'$|| \nabla \mathbf{p} ||^2_2$')
ax[2].set_ylabel(r'$ \frac{ \partial^2 || \mathbf{G p} - \mathbf{d} ||_2^2 } { {|| \nabla \mathbf{p} ||^2_2}^2} $')
```

We can clearly see that the L-curve finds $\lambda$ of about 1e-4 to 1e-4, which, which is consistent with what is obtained by the second derivative approahc. However, from the left curve it appears that the optimum $\lambda$ is actually close to 1, which shows that in this example the value estimated by the L-curve is far off. Below we invert the load distribution using either the optimum $\lambda$ or the one derived with the L-curve. 

```python
ps_laplacian_linear = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='linear', lamb=lambda_list_laplacian[idx_lambda_Lcurve_laplacian],Cm1=laplacian,G=G)
ps_laplacian_lbfgs = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='lbfgs', lamb=lambda_list_laplacian[idx_lambda_Lcurve_laplacian],Cm1=laplacian,G=G,
                                         epsilon=1e-35,maxit=3e2)
ps_laplacian_lbfgs_fast = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='lbfgs_fast', lamb=lambda_list_laplacian[idx_lambda_Lcurve_laplacian],Cm1=laplacian,G=G,
                                         epsilon=1e-35,maxit=1e4)
#for the constraint we add the intel that we know that the load is 0 everywhere around the lake.
constraint = ps.copy()
constraint[np.where(constraint!=0)] = np.nan #setting to nan the value of the load of the region corresponding to the lake.
ps_laplacian_lbfgs_constrained = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='lbfgs', lamb=lambda_list_laplacian[idx_lambda_Lcurve_laplacian],Cm1=laplacian,G=G,
                                         epsilon=1e-35,maxit=3e2,constraint=constraint)
ps_laplacian_lbfgs_fast_constrained = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='lbfgs_fast', lamb=lambda_list_laplacian[idx_lambda_Lcurve_laplacian],Cm1=laplacian,G=G,
                                         epsilon=1e-35,maxit=1e4,constraint=constraint)

#### TELLING WHICH INVERTED LOAD IS THE BEST #### 
print(r'############## $\lambda$ derived from Lcurve ###############')
print(r'The error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ with Linear inversion is :',np.linalg.norm(ps-ps_laplacian_linear))
print(r"The error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ with our unconstrained LBFGS implementation is :",np.linalg.norm(ps-ps_laplacian_lbfgs))
print(r"The error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ with scipy's unconstrained LBFGS implementation is :",np.linalg.norm(ps-ps_laplacian_lbfgs_fast))
print(r"The error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ with our constrained LBFGS implementation is :",np.linalg.norm(ps-ps_laplacian_lbfgs_constrained))
print(r"The error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ with scipy's constrained LBFGS implementation is :",np.linalg.norm(ps-ps_laplacian_lbfgs_fast_constrained))
print('##############################################"')
#### plottign
fig, ax = plt.subplots(2,3,figsize=(13,6))
fig.suptitle(r'Inversion with $\lambda$ derived from L-curve', fontsize=16)
vmin = np.min([ps_laplacian_linear,ps_laplacian_lbfgs,ps_laplacian_lbfgs_fast,ps_laplacian_lbfgs_fast_constrained,ps_laplacian_lbfgs_constrained])
vmax = np.max([ps_laplacian_linear,ps_laplacian_lbfgs,ps_laplacian_lbfgs_fast,ps_laplacian_lbfgs_fast_constrained,ps_laplacian_lbfgs_constrained])
####
im = ax[0,0].pcolormesh(lonr,latr,ps_laplacian_linear, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0,0].set_ylabel('latitude (°)')
ax[0,0].set_xlabel('longitude (°)')
ax[0,0].set_title('linear no constraint')
####
im = ax[0,1].pcolormesh(lonr,latr,ps_laplacian_lbfgs, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0,1].set_ylabel('latitude (°)')
ax[0,1].set_xlabel('longitude (°)')
ax[0,1].set_title('lbfgs no constraint')
####
im = ax[0,2].pcolormesh(lonr,latr,ps_laplacian_lbfgs_fast, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0,2].set_ylabel('latitude (°)')
ax[0,2].set_xlabel('longitude (°)')
ax[0,2].set_title('lbfgs_fast no constraint')
####
im = ax[1,0].pcolormesh(lonr,latr,constraint, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[1,0].set_ylabel('latitude (°)')
ax[1,0].set_xlabel('longitude (°)')
ax[1,0].set_title('constraint')
###
im = ax[1,1].pcolormesh(lonr,latr,ps_laplacian_lbfgs_constrained, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[1,1].set_ylabel('latitude (°)')
ax[1,1].set_xlabel('longitude (°)')
ax[1,1].set_title('lbfgs with constraint')
# ####
im = ax[1,2].pcolormesh(lonr,latr,ps_laplacian_lbfgs_fast_constrained, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[1,2].set_ylabel('latitude (°)')
ax[1,2].set_xlabel('longitude (°)')
ax[1,2].set_title('lfbsg fast with constraint')
fig.subplots_adjust(wspace=0.4, hspace=0.4)
####

ps_laplacian_linear = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='linear', lamb=lambs[idx_lambda_optimum_laplacian],Cm1=laplacian,G=G)
ps_laplacian_lbfgs = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='lbfgs', lamb=lambs[idx_lambda_optimum_laplacian],Cm1=laplacian,G=G,
                                         epsilon=1e-35,maxit=3e2)
ps_laplacian_lbfgs_fast = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='lbfgs_fast', lamb=lambs[idx_lambda_optimum_laplacian],Cm1=laplacian,G=G,
                                         epsilon=1e-35,maxit=1e4)
ps_laplacian_lbfgs_constrained = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='lbfgs', lamb=lambs[idx_lambda_optimum_laplacian],Cm1=laplacian,G=G,
                                         epsilon=1e-35,maxit=3e2,constraint=constraint)
ps_laplacian_lbfgs_fast_constrained = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='lbfgs_fast', lamb=lambs[idx_lambda_optimum_laplacian],Cm1=laplacian,G=G,
                                         epsilon=1e-35,maxit=1e4,constraint=constraint)

print(r'############## $\lambda$ optimum ###############')
print(r'The error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ with Linear inversion is :',np.linalg.norm(ps-ps_laplacian_linear))
print(r"The error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ with our unconstrained LBFGS implementation is :",np.linalg.norm(ps-ps_laplacian_lbfgs))
print(r"The error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ with scipy's unconstrained LBFGS implementation is :",np.linalg.norm(ps-ps_laplacian_lbfgs_fast))
print(r"The error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ with our constrained LBFGS implementation is :",np.linalg.norm(ps-ps_laplacian_lbfgs_constrained))
print(r"The error $|| \mathbf{p}_{true} - \mathbf{p}_{cal} ||_2^2$ with scipy's constrained LBFGS implementation is :",np.linalg.norm(ps-ps_laplacian_lbfgs_fast_constrained))
print('##############################################"')

##plotting 
fig, ax = plt.subplots(2,3,figsize=(13,6))
fig.suptitle(r'Inversion with optimum $\lambda$', fontsize=16)
vmin = np.min([ps_laplacian_linear,ps_laplacian_lbfgs,ps_laplacian_lbfgs_fast,ps_laplacian_lbfgs_fast_constrained,ps_laplacian_lbfgs_constrained])
vmax = np.max([ps_laplacian_linear,ps_laplacian_lbfgs,ps_laplacian_lbfgs_fast,ps_laplacian_lbfgs_fast_constrained,ps_laplacian_lbfgs_constrained])
####
im = ax[0,0].pcolormesh(lonr,latr,ps_laplacian_linear, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0,0].set_ylabel('latitude (°)')
ax[0,0].set_xlabel('longitude (°)')
ax[0,0].set_title('linear no constraint')
####
im = ax[0,1].pcolormesh(lonr,latr,ps_laplacian_lbfgs, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0,1].set_ylabel('latitude (°)')
ax[0,1].set_xlabel('longitude (°)')
ax[0,1].set_title('lbfgs no constraint')
####
im = ax[0,2].pcolormesh(lonr,latr,ps_laplacian_lbfgs_fast, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[0,2].set_ylabel('latitude (°)')
ax[0,2].set_xlabel('longitude (°)')
ax[0,2].set_title('lbfgs_fast no constraint')
####
im = ax[1,0].pcolormesh(lonr,latr,constraint, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[1,0].set_ylabel('latitude (°)')
ax[1,0].set_xlabel('longitude (°)')
ax[1,0].set_title('constraint')
###
im = ax[1,1].pcolormesh(lonr,latr,ps_laplacian_lbfgs_constrained, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[1,1].set_ylabel('latitude (°)')
ax[1,1].set_xlabel('longitude (°)')
ax[1,1].set_title('lbfgs with constraint')
# ####
im = ax[1,2].pcolormesh(lonr,latr,ps_laplacian_lbfgs_fast_constrained, vmin=vmin, vmax=vmax)
cbar = plt.colorbar(im)
cbar.set_label('load (Pa)')
ax[1,2].set_ylabel('latitude (°)')
ax[1,2].set_xlabel('longitude (°)')
ax[0,2].set_title('lfbsg fast with constraint')
fig.subplots_adjust(wspace=0.4, hspace=0.4)
```

## Inversion with complex load distribution and small noise 

We have seen that with the Laplacian approach the L-curve is was not able to crrectly estimate the value of $\lambda$, and therefore we had to resort to an empirical criterion. We also saw that the Gaussian regularization was a potential alternative to the Laplacian, even though the L-curve seemed less effective to estimate the correct value of $\lambda$ with this approach. In this section, we will therefore investigate wheter or not an L-curve performed with the gaussian regularization can actually estimate the correct value of $\lambda$.

```python
#now we perform the inversion with linear and LBFGS approahces and with and without constraint
lambs = np.logspace(-10,3,50)
sigmas = np.power(10, np.arange(-9., 10.)) 
sigmas = sigmas[7:-7] #réduit le nombre de sigmas pour aller plus vite

gaussians = np.zeros((source_number,source_number,len(sigmas)))

for i in range(len(sigmas)):
    gaussians[:,:,i] = np.load(f'gaussian_salt_{sigmas[i]}.npy')

norm_gaussian_ps = np.zeros((len(sigmas),len(lambs)))
norm_residuals_gaussian = np.zeros((len(sigmas),len(lambs)))
norm_diff_ps_gaussian = np.zeros((len(sigmas),len(lambs))) 
misfit_horizontal = np.zeros((len(sigmas),len(lambs)))
misfit_vertical = np.zeros((len(sigmas),len(lambs)))

disp_obs = Us_multiple_noisy.reshape((data_number*3,1)) #in this case we set data number as the number of stations instead of the actual number of data

print('ok')
for i in trange(len(lambs)):
    lamb = lambs[i]
    for j, sigma in enumerate(sigmas):
        gaussian = gaussians[:,:,j]
        ps_gaussian = disp2load.disp2load(E,v,rs,xs,ys,Us_multiple_noisy,mode='linear', lamb=lamb,Cm1=gaussian,G=G) #precomputed G and lalaplcian to save time
        print('done')
        ps_gaussian_reshaped = ps_laplacian.reshape((source_number,1))
        norm_gaussian_ps[j,i] = np.sqrt(np.sum((gaussian@ps_gaussian_reshaped)**2)) 
        disp_cal = G@ps_gaussian_reshaped
        norm_residuals_gaussian[j,i] = np.sqrt(np.sum((disp_obs-disp_cal)**2))
        norm_diff_ps_gaussian[j,i] =  np.sqrt(np.sum((ps-ps_gaussian)**2))    
        
        Us_cal = disp_cal.reshape((1700,3)) #reshaping in the same shape as the Us_multiple_noisy  to more easily separate the displacements on the diff components 
        diff_e = np.mean(np.abs(Us_cal[:,0]-Us_multiple_noisy[:,0]))
        diff_n = np.mean(np.abs(Us_cal[:,1]-Us_multiple_noisy[:,0]))
        misfit_horizontal[j,i] = np.mean([diff_e, diff_n]) 
        misfit_vertical[j,i] = np.mean(np.abs(Us_cal[:,2]-Us_multiple_noisy[:,2]))
```

## Difficulties

-> strong wolfe conditions fail to converge sometimes -> so need to use armijo when it does, need to give an estimate of how many interations need to fail to consider that the strong wolfe conditons are indeed dailing -> based myself on the number that is used in scipy implemetntation they also use 20 fails to consider that conditons are failing 

-> the curvature condition cannot be met so wolfe does not converge, and therefore I use the Armijo linesearch, which only respects the sufficient decrease conditions, which of course degrades the performance of the BFGS/L-BFGS algorithm.
