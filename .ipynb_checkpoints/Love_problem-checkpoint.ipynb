```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import tqdm, trange
import load2disp
import utils
from multiprocessing import Pool
import time
import concurrent.futures
```

```python
def lat_lng_to_cartesian(lat, lng):
    """Converts lat lon coordinates to cartesian"""
    EARTH_RADIUS = 6371000
    # Convert latitude and longitude from degrees to radians
    lat_rad = np.deg2rad(lat)
    lng_rad = np.deg2rad(lng)
    # Calculate Cartesian coordinates
    x = EARTH_RADIUS * np.cos(lat_rad) * np.cos(lng_rad)
    y = EARTH_RADIUS * np.cos(lat_rad) * np.sin(lng_rad)
    z = EARTH_RADIUS * np.sin(lat_rad)
    return x, y, z

def cartesian_to_lat_lng(x, y, z):
    """Converts cartesian coordinates to lat lon"""
    EARTH_RADIUS = 6371000
    # Calculate latitude in radians
    lat_rad = np.arctan2(z, np.sqrt(x**2 + y**2))
    # Calculate longitude in radians
    lng_rad = np.arctan2(y, x)
    # Convert latitude and longitude from radians to degrees
    lat = np.rad2deg(lat_rad)
    lng = np.rad2deg(lng_rad)
    return lat, lng
```

```python
#################### WHERE THE DISPLACEMENT IS COMPUTED #################"
#providing boundaries and ticks for the general domain  ->
xs = np.linspace(-500, 500, 40)
ys = np.linspace(-500, 500, 40)
zs = np.linspace(0,100,40) #fait de la 3D! 
Us = np.zeros((len(ys), len(xs), len(zs),3)) #cr la map des dplacement initialise  0 

##################################### DEFINING SOURCES WITH THEIR PARAMETERS #############
ps = np.random.randint(low=0, high=9.81*10*1000, size=(2,2))  
# ps = np.ones((2,2))*9.81*10*1000
E = 20e9
v = 0.25
#creating a mesh wtih all sources (100 sources in our case where we use ps of size (10*10)) 
rs = utils.create_source_mesh(-450,450,-450,450,ps) #sources should have part out of the domain in which we compute 

############################ LOOPING OVER SOURCES (SOURCES STORED IN 2D ARRAY TO BE SIMILAR TO GRIDDING OF PRESSURE)######
tim = time.time()
for vx_idx in range(len(ps[0,:])): #boucle sur les sources et on va sommet les rusltats sur les vectices pour faire le dlacement gloabel 
    for vy_idx in range(len(ps[:,0])): 
        p = ps[vy_idx,vx_idx]
        r = rs[vy_idx,vx_idx]
        
        with concurrent.futures.ProcessPoolExecutor() as executor:
            futures = []
            #computing the combinations of coordinate points where the displacement will have to be evaluated
            for i, x in enumerate(xs):
                for j, y in enumerate(ys):
                    for k, z in enumerate(zs):
                        xyz = [x, y, z]
                        futures.append(executor.submit(load2disp.load2disp, xyz, r, p, E, v))

            ######### now finally computing the displacement and putting it in Us 
            for i, x in enumerate(xs):
                for j, y in enumerate(ys):
                    for k, z in enumerate(zs):
                        U = futures.pop(0).result()
                        Us[j, i,k,:] += U.reshape(3)  #we add up the contribution of sources

print('It took',time.time()-tim,f's to compute the displacement at {Us.shape[0]*Us.shape[1]*Us.shape[2]} points for {ps.shape[0]*ps.shape[1]} sources ') #
```

Maintenant on va calculer le strain tensor en tout point !! (sauf aux bords -> peut tre faire euler forward en z pour pouvoir calculer dformation  la surface...)

## Strain tensor computation

```python
dx = xs[1]-xs[0]
dy = ys[1]-ys[0]
dz = zs[1]-zs[0]

strain_matrix = np.zeros((len(ys),len(xs),len(zs),3,3)) #on pourra pas calculer le strain dans les points aux bords avec notre implmentation -> pourrait faire euler forward sii on veut pouvoir avoir 
#la dformation  la surface... 

############################ LOOPING OVER SOURCES (SOUCRES STORED IN 2D ARRAY TO BE SIMILAR TO GRIDDING OF PRESSURE)######
tim = time.time()

with concurrent.futures.ProcessPoolExecutor() as executor:
    futures = []
    #computing the different combinations of Up to feed in order to compute the strain at each point -> chaque point on feed les composantes de U aux points suivant et prcdent selon les 3 composantes = 6 valeurs  donner
    for i in range(1,len(xs)-1):
        for j in range(1,len(ys)-1):
            for k in range(1,len(zs)-1):
                Up = np.array([[Us[j,i-1,k,0],Us[j,i+1,k,0]], #ligne ou on met les dplacement dans direction x 
                               [Us[j-1,i,k,1],Us[j+1,i,k,1]], #ligne dplacement direction y
                               [Us[j,i,k-1,2],Us[j,i,k+1,2]]]) #ligne dplacent en z 
                futures.append(executor.submit(utils.compute_strain_tensor_at_point,Up,dx,dy,dz))

    ######### now finally computing the displacement and putting it in Us 
    for i in range(1,len(xs)-1):
        for j in range(1,len(ys)-1):
            for k in range(1,len(zs)-1):
                strain = futures.pop(0).result()
                strain_matrix[j,i,k,:] = strain.reshape(3,3)  

print('Took ', time.time()-tim, f's to compute {strain_matrix.shape[0]*strain_matrix.shape[1]*strain_matrix.shape[2]} strain tensors') #
```

```python
plt.figure()
plt.title('Dformation verticale (epsilon zz)    z = deuxime indice   ')
plt.pcolormesh(ys,xs,strain_matrix[:,:,1,2,2]) #affiche la carte horizontale  l'index 1 (premier index ou calcul dans la direction z)  de epsilon zz 
plt.colorbar()
#on voit que la dformation est maximale en priphrique du bloc  (ngative ou positive )  !!! 

plt.figure()
plt.title('Dformation horizontale dans direction de x (epsilon xx)    z = deuxime indice   ')
plt.pcolormesh(ys,xs,strain_matrix[:,:,1,0,0]) #affiche la carte horizontale  l'index 1 (premier index ou calcul dans la direction z)  de epsilon zz 
plt.colorbar() #dplacement en x fait bien vers la droite  gauche dt vers la gauche  droite  du coup a fait forte slope dans le dplacement au milieu et par contre plateau sur 
#cts du coup quand fait drive selon x,  a fait valeur ngative constante au milieu + 0 sur les bords 

```

Maintenant que l'on a les tenseurs de contrainte en tout point, il nous suffit de calculer d'utiliser l'hypothse lastique et on fait x E pour avoir sigma:

## Stress tensor computation

```python
E = 1 # 20e9
stress_matrix = E*strain_matrix
print(stress_matrix.shape) #shape y*x*z * tensor
```

Maintenant que l'on a tous les stress tensors, il faudrait tre capable de placer la faille dans le mesh pour ensuite faire la rotation rotation des tenseurs de de contrainte et en appliquer la formule pour calculer le stress de coulomb associ 

## Defining the fault and matrixrotation

```python
#on dfini la faille comme un plan vertical pour le moment -> on connait ainsi aisment o
#des pistes pour une gnralisation avec des points discrets pour modliser la faille ont dj t inities sur le jupyter hub 

#coordonnes de la faille 
xf = [200,420]
yf = [200,300] 

#calcul de l'azimuth de la daille
az = np.arctan((xf[1]-xf[0])/(yf[1]-yf[0]))
az = np.pi/2

# rotation d'un tenseur de contrainte :
rot_z = np.array([[np.cos(-az), -np.sin(-az), 0],
                   [np.sin(-az), np.cos(-az), 0],
                    [0, 0, 1]])

# Reshape the large_array to (40*40*40, 3, 3)
reshaped_stress_matrix = stress_matrix.reshape(-1, 3, 3)

# Perform matrix multiplication between the 3x3 matrix and all matrices in the reshaped array
result = np.matmul(rot_z,reshaped_stress_matrix)  # voir si a a bien fait 

# Reshape the result back to the original shape (40, 40, 40, 3, 3)
rotated_stress_matrix = result.reshape(40, 40, 40, 3, 3)


plt.figure()
plt.title('stress.. ')
plt.pcolormesh(ys,xs,rotated_stress_matrix[:,:,1,2,2])
plt.plot(xf,yf,color='red')
plt.colorbar()
```

Maintenant que dait la rotation, la direction x devrait tre la direction normale au plan et les directions z et y devraient tre tangentes au plan. On devrait ainsi  prsent pouvoir calculer le stress de coulomb 

## Coulomb stress computation

```python
#calcul du stress de coulomb  -> pusique on a  ortated  xx est normal et xy et xz est 

friction_coeff = 1

coulomb_sress_matrix = rotated_stress_matrix[:,:,:,0,1] +friction_coeff*rotated_stress_matrix[:,:,:,0,0]

print(coulomb_sress_matrix.shape)

plt.figure()
plt.title('coulomb stress.. ')
plt.pcolormesh(ys,xs,coulomb_sress_matrix[:,:,1])
plt.plot(xf,yf,color='red')
plt.colorbar()
```
